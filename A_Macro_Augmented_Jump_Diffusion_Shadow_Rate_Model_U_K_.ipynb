```python
!pip -q install filterpy==1.4.5 scikit-learn==1.4.2 statsmodels==0.14.1

import numpy as np, pandas as pd, matplotlib.pyplot as plt
from pathlib import Path; from scipy.optimize import minimize
from filterpy.kalman import KalmanFilter
from sklearn.decomposition import PCA; from sklearn.preprocessing import StandardScaler
from statsmodels.regression.linear_model import OLS; from statsmodels.tools.tools import add_constant
from statsmodels.tsa.stattools import adfuller
from statsmodels.stats.diagnostic import acorr_ljungbox, het_arch
from statsmodels.tsa.ar_model import AutoReg
from IPython.display import display

SEED = 42
rng  = np.random.default_rng(SEED);  np.random.seed(SEED)

DATA = Path("/content/data shadow new.xlsx")
xls  = pd.ExcelFile(DATA)

def _prep(sheet, cols):
    df = pd.read_excel(xls, sheet, header=1)
    df.columns = cols
    df["Date"] = pd.to_datetime(df["Date"], dayfirst=True, errors="coerce")
    return df.dropna(subset=["Date"]).set_index("Date")

ser = {
    "SR" : _prep("Shadow Rates UK", ["Date","SR UK"]),
    "INF": _prep("Inflation Data UK", ["Date","INF UK"]),
    "UN" : _prep("Unemployment Data UK", ["Date","UNEM UK"]),
    "CDS": _prep("Sovereign CDS Spreads UK", ["Date","CDS UK"]),
    "GDP": _prep("Real GDP UK", ["Date","GDP UK"]),
    "EX" : _prep("Exchange Rates UK", ["Date","EX UK"]),
    "3M" : _prep("3M",  ["Date","3M UK"]),
    "6M" : _prep("6M",  ["Date","6M UK"]),
    "1Y" : _prep("1Y",  ["Date","1Y UK"]),
    "2Y" : _prep("2Y",  ["Date","2Y UK"]),
    "5Y" : _prep("5Y",  ["Date","5Y UK"]),
    "7Y" : _prep("7Y",  ["Date","7Y UK"]),
    "10Y": _prep("10Y", ["Date","10Y UK"]),
    "30Y": _prep("30Y", ["Date","30Y UK"]),
}

df = pd.concat(ser.values(), axis=1).interpolate("linear")
macro = ["INF UK","UNEM UK","EX UK","CDS UK","GDP UK"]
ylds  = ["3M UK","6M UK","1Y UK","2Y UK","5Y UK","7Y UK","10Y UK","30Y UK"]
df[ylds] = df[ylds] * 100
df    = df.dropna(subset=["SR UK"] + macro + ylds)

y_df, m_df = df[ylds].copy(), df[macro].copy()

var_list = ylds + macro
tbl_station, tbl_ac, tbl_arch = [], [], []

for v in var_list:
    s0 = df[v].dropna()
    adf0 = adfuller(s0, autolag='AIC'); stat0, p0 = adf0[0], adf0[1]

    if p0 < .05:
        tbl_station.append([v, stat0, p0, 'Stat']);  s_final = s0
    else:
        s1 = s0.diff().dropna()
        adf1 = adfuller(s1, autolag='AIC')
        tbl_station += [[f'{v}(orig)',stat0,p0,'NOT stat'],
                        [f'{v}(Δ1)',adf1[0],adf1[1],
                         'Stat' if adf1[1]<.05 else 'NOT']]
        s_final = s1

    lb_p   = acorr_ljungbox(s_final, lags=[10], return_df=True)['lb_pvalue'].iloc[0]
    arch_p = het_arch(s_final, maxlag=10)[1]
    tbl_ac.append([v, lb_p,   'Sig' if lb_p<.05  else 'OK'])
    tbl_arch.append([v, arch_p,'Sig' if arch_p<.05 else 'OK'])

print("\n=== ADF ===");   display(pd.DataFrame(tbl_station, columns=['Var','ADF','p','Res']))
print("\n=== LB  ===");  display(pd.DataFrame(tbl_ac,      columns=['Var','p','AC?']))
print("\n=== ARCH ==="); display(pd.DataFrame(tbl_arch,    columns=['Var','p','ARCH?']))

if y_df.abs().max().max() > 20: y_df /= 100
low, upp = y_df.mean()-3*y_df.std(), y_df.mean()+3*y_df.std()
y_df = y_df.clip(lower=low, upper=upp, axis=1)
m_z  = pd.DataFrame(StandardScaler().fit_transform(m_df),
                    index=m_df.index, columns=m_df.columns)

def _beta(y,z):
    X = add_constant(z); return {c: OLS(y[c],X).fit().params.values for c in y}

def _pca(y): return PCA(1,random_state=SEED).fit_transform(y).ravel()

def _sim(n,r0,z,β,ψ,lb=-2):
    κ,σ,λ,μ,s,θ0,θ1,θ2 = ψ; z0,z1 = 240,1992
    S=np.empty(n); S[0]=r0
    for t in range(1,n):
        θ = θ0 if t<z0 else (θ1 if t<=z1 else θ2)
        S[t]=S[t-1]+κ*(θ-S[t-1])+σ*rng.normal()
        if rng.random()<λ: S[t]+=rng.normal(μ,s)
        if S[t]<lb: S[t]=lb+0.3*(S[t]-lb)
        Xt=z.iloc[t].values
        S[t]+=0.5*np.mean([Xt@β[m][1:] for m in β])
    return S

def _kalman(x):
    kf=KalmanFilter(dim_x=1,dim_z=1)
    kf.x=[[x[0]]]; kf.F=kf.H=np.eye(1); kf.P*=100; kf.R[:]=1; kf.Q[:]=0.01
    out=[]
    for v in x: kf.predict(); kf.update([[v]]); out.append(kf.x[0,0])
    return np.array(out)

β   = _beta(y_df,m_z); pca = _pca(y_df)
print("\n=== β̂m (macro loading per maturity) ===")
for m, coef in β.items():
    print(f"{m:>5}: {coef[1]:+.4f}")
pca = _pca(y_df)
θ0, θ1, θ2 = pca[:50].mean(), pca[50:150].mean(), pca[150:].mean()

θ0,θ1,θ2 = pca[:50].mean(), pca[50:150].mean(), pca[150:].mean()
x0  = [0.15,0.015,0.005,0.10,0.05, θ0,θ1,θ2]
bnd = [(1e-4,1),(1e-4,0.05),(0,0.1),(-.5,.5),(1e-4,0.2),(-5,5),(-5,5),(-5,5)]
obj = lambda p: ((_sim(len(pca),pca[0],m_z,β,p)-pca)**2).sum()
ψhat= minimize(obj,x0,bounds=bnd,method='L-BFGS-B',
               options={'maxiter':500}).x

S     = _sim(len(pca),pca[0],m_z,β,ψhat)
S_sm  = _kalman(S)
final = pd.Series(0.7*pca+0.3*S_sm,index=df.index,name='Final SR')

print("\nψ̂:",np.round(ψhat,4))
display(final.head())

ws   = np.arange(0.50, 0.95, 0.05)
rmse = []

for w in ws:
    shadow = w*pca + (1-w)*S_sm
    errs = []
    for m in ylds:
        y_hat = OLS(y_df[m], add_constant(shadow)).fit().fittedvalues
        errs.append(((y_df[m]-y_hat)**2).mean())
    rmse.append(np.sqrt(np.mean(errs))*1e4)

grid = pd.DataFrame({"w": ws, "RMSE_bp": rmse})
print(grid.round(2))

grid.plot(x="w", y="RMSE_bp", marker="o", figsize=(4,3.5))
plt.ylabel("mean cross-section RMSE (bp) (U.K.)"); plt.tight_layout(); plt.show()

plt.figure(figsize=(5,3.5))
plt.bar(range(1,9), PCA().fit(y_df).explained_variance_ratio_)
plt.xlabel('PC'); plt.ylabel('Explained var share')
plt.title('Scree Plot – United Kingdom Yield Curve'); plt.tight_layout(); plt.show()

plt.figure(figsize=(11,5))
plt.plot(df.index,final,label='Final SR')
plt.plot(df.index,pca,'--',label='PCA'); plt.grid(); plt.legend()
plt.title('UK shadow-rate'); plt.show()

start,end='2009-01-01','2022-08-01'; alpha=0.05
ewma = final.ewm(alpha=alpha,adjust=False).mean().loc[start:end]
wu   = df['SR UK'].loc[start:end]

plt.figure(figsize=(11,5))
plt.plot(final.resample('M').mean().loc[start:end],lw=2,label='MAJD')
plt.plot(wu.resample('M').last(),'--',label='WX'); plt.grid(); plt.legend()
plt.ylabel('%'); plt.title('MAJD vs WX (United Kingdom)'); plt.show()

plt.figure(figsize=(11,5))
plt.plot(ewma,lw=2,label=f'Final EWMA (α={alpha})')
plt.plot(wu,'--',label='Wu-Xia'); plt.grid(); plt.legend()
plt.ylabel('%'); plt.title('Final EWMA vs Wu-Xia (daily)'); plt.show()

fit_tbl=[]
for m in ylds:
    rmse_f = np.sqrt(((OLS(y_df[m], add_constant(final)).fit()
                       .fittedvalues - y_df[m])**2).mean())*1e4
    rmse_w = np.sqrt(((OLS(y_df[m], add_constant(df['SR UK'])).fit()
                       .fittedvalues - y_df[m])**2).mean())*1e4
    fit_tbl.append([m, rmse_f, rmse_w])

fit_df = pd.DataFrame(fit_tbl, columns=['Maturity','RMSE_final','RMSE_WX'])
print("\n=== In-sample cross-section RMSE (bp) ==="); display(fit_df)

TEST_START = '2023-01-01'
train      = final.loc[:TEST_START]
test_idx   = final.loc[TEST_START:].index
n_train    = len(train)

oos_len   = len(test_idx)

ar1 = AutoReg(train, lags=1, old_names=False).fit()

pred_path = ar1.forecast(steps=oos_len)
pred_path.index = test_idx

fc_df = pd.DataFrame({f'h{h}': pred_path.shift(h-1)
                      for h in range(1, 13)},
                     index=test_idx)

oos_tbl=[]
for m in ylds:
    a,b = OLS(y_df[m].loc[:TEST_START], add_constant(train)).fit().params
    act = y_df[m].loc[TEST_START:]
    for h in range(1,13):
        rmse = np.sqrt(((a + b*fc_df[f'h{h}'] - act)**2).mean())*1e4
        oos_tbl.append([m, h, rmse])

oos_df = (pd.DataFrame(oos_tbl, columns=['Maturity','h','RMSE_bp'])
            .pivot(index='Maturity', columns='h', values='RMSE_bp')
            .reindex(ylds))

print("\n=== OOS RMSE (bp) – forecast horizon h = 1…12 months ===")
display(oos_df.round(2))

OUT = Path("/content/shadow_rate_uk.csv")
final.to_csv(OUT, index_label="Date")
final.to_excel("/content/shadow_rate_uk.xlsx", sheet_name="UK SR")

print(f"\n✓ Shadow-rate series written to:  {OUT.resolve()}")
print(f"  Rows written: {len(final):,}")

import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from statsmodels.tsa.api import VAR
import matplotlib.pyplot as plt

file_path = '/content/favar uk.xlsx'
wu_xia_file_path = '/content/wu xia shadow rates.xlsx'

wu_xia_data = pd.ExcelFile(wu_xia_file_path)
wu_xia_uk = pd.read_excel(wu_xia_data, sheet_name='Shadow Rates UK', skiprows=1)
wu_xia_uk.columns = ['Date', 'WX (United Kingdom)']
wu_xia_uk['Date'] = pd.to_datetime(wu_xia_uk['Date'], errors='coerce')
wu_xia_uk = wu_xia_uk.dropna(subset=['Date']).set_index('Date')

data = pd.ExcelFile(file_path)
shadow_rates_uk = pd.read_excel(data, sheet_name='Shadow Rates UK', index_col='Date', parse_dates=True)
industrial_prod_uk = pd.read_excel(data, sheet_name='Industrial Production UK', index_col='Date', parse_dates=True)
cpi_uk = pd.read_excel(data, sheet_name='CPI UK', index_col='Date', parse_dates=True)
unemployment_uk = pd.read_excel(data, sheet_name='Unemployment UK', index_col='Date', parse_dates=True)
bond_yields_uk_3m = pd.read_excel(data, sheet_name='Bond Yields UK 3M', index_col='Date', parse_dates=True)
bond_yields_uk_10y = pd.read_excel(data, sheet_name='Bond Yields UK 10Y', index_col='Date', parse_dates=True)

uk_data_with_wu_xia = pd.concat([
    wu_xia_uk,
    shadow_rates_uk.rename(columns={'Shadow Rate UK': 'MAJD (United Kingdom)'}),
    industrial_prod_uk,
    cpi_uk,
    unemployment_uk,
    bond_yields_uk_3m,
    bond_yields_uk_10y
], axis=1).dropna()

print(uk_data_with_wu_xia.columns)

SHADOW_COLS_UK = ['WX (United Kingdom)', 'MAJD (United Kingdom)']
FACTOR_COLS_UK = [c for c in uk_data_with_wu_xia.columns if c not in SHADOW_COLS_UK]

diff_uk = uk_data_with_wu_xia.diff().dropna()
diff_uk.index = pd.date_range(start=diff_uk.index[0], periods=len(diff_uk), freq='ME')

standardized_factor_data_uk = (
    diff_uk[FACTOR_COLS_UK] - diff_uk[FACTOR_COLS_UK].mean()
) / diff_uk[FACTOR_COLS_UK].std()

pca_uk = PCA(n_components=3)
pca_factors_uk = pca_uk.fit_transform(standardized_factor_data_uk)

common_factors_uk = pd.DataFrame(
    pca_factors_uk,
    index=diff_uk.index,
    columns=['Factor1', 'Factor2', 'Factor3']
)

def process_data_for_VAR(diff_data, common_factors, shadow_rate_col):
    var_data = pd.concat([common_factors, diff_data[[shadow_rate_col]]], axis=1).dropna()
    return var_data

uk_favar_data_your_shadow = process_data_for_VAR(
    diff_uk,
    common_factors_uk,
    shadow_rate_col='MAJD (United Kingdom)'
)

uk_favar_data_wu_xia_shadow = process_data_for_VAR(
    diff_uk,
    common_factors_uk,
    shadow_rate_col='WX (United Kingdom)'
)

def fit_VAR_and_analyze(data, lags=2, n_periods=10):
    """
    • Fits a VAR(lags) to the input DataFrame.
    • Prints the full coefficient tables, information criteria,
      and residual-correlation matrix.
    • Returns both the fitted VAR object and its orthogonalised IRF object.
    """
    model     = VAR(data)
    var_model = model.fit(lags)

    print("\n=== VAR({}) estimation results ===".format(lags))
    print(var_model.summary())
    print("\nCorrelation matrix of residuals")
    print(var_model.resid.corr())

    irf = var_model.irf(n_periods)
    return var_model, irf

var_your_shadow, irf_your_shadow = fit_VAR_and_analyze(
    uk_favar_data_your_shadow, lags=2, n_periods=10)

var_wu_xia, irf_wu_xia_shadow = fit_VAR_and_analyze(
    uk_favar_data_wu_xia_shadow, lags=2, n_periods=10)

print("Impulse Response for the MAJD (United Kingdom):")
fig = irf_your_shadow.plot(orth=True)
for ax in fig.axes:
    ax.set_title(ax.get_title().replace("MAJD (United Kingdom)", "MAJD (U.K.)"))
plt.show()

print("Impulse Response for WX (United Kingdom):")
fig = irf_wu_xia_shadow.plot(orth=True)
for ax in fig.axes:
    ax.set_title(ax.get_title().replace("WX (United Kingdom)", "WX (U.K.)"))
plt.show()

def calculate_cumulative_irf(irf_object, n_periods=10):
    cumulative_irf = {}
    irf_matrix = irf_object.orth_irfs[:n_periods]
    response_names = irf_object.model.names

    for i, impulse in enumerate(response_names):
        for j, response in enumerate(response_names):
            response_values = irf_matrix[:, j, i]
            cumulative_irf[f"{response} → {impulse}"] = np.sum(np.abs(response_values))

    return cumulative_irf

cumulative_irf_your_shadow = calculate_cumulative_irf(irf_your_shadow, n_periods=10)
cumulative_irf_wu_xia_shadow = calculate_cumulative_irf(irf_wu_xia_shadow, n_periods=10)

comparison_df = pd.DataFrame({
    "MAJD (United Kingdom)": cumulative_irf_your_shadow,
    "WX (United Kingdom)": cumulative_irf_wu_xia_shadow
}).T

comparison_df.loc['Difference'] = (
    comparison_df.loc['MAJD (United Kingdom)'] - comparison_df.loc['WX (United Kingdom)']
)

print("Cumulative Impulse Responses Comparison for UK:")
print(comparison_df)

output_file = 'cumulative_irf_comparison_uk.csv'
comparison_df.to_csv(output_file)
print(f"Comparison data saved as '{output_file}' in the current working directory.")

comparison_df.loc["MAJD (United Kingdom)"].plot(kind="bar", figsize=(10, 6), alpha=0.7, label="MAJD")
comparison_df.loc["WX (United Kingdom)"].plot(kind="bar", figsize=(10, 6), alpha=0.7, label="WX", color="orange")
plt.title("Cumulative Impulse Responses Comparison (United Kingdom)")
plt.xlabel("Response → Impulse")
plt.ylabel("Cumulative IRF")
plt.legend()
plt.tight_layout()
plt.show()

!pip install statsmodels

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.api import VAR

file_path = '/content/new entry uk.xlsx'
xls = pd.ExcelFile(file_path)

shadow_uk = pd.read_excel(xls, sheet_name='My shadow rates UK', parse_dates=['Date'])
inflation_uk = pd.read_excel(xls, sheet_name='Inflation UK', parse_dates=['Date'])
gdp_uk = pd.read_excel(xls, sheet_name='Real GDP UK', parse_dates=['Date'])

shadow_uk.set_index('Date', inplace=True)
if len(shadow_uk.columns) == 1:
    shadow_uk.columns = ['MAJD (United Kingdom)']

inflation_uk.set_index('Date', inplace=True)
gdp_uk.set_index('Date', inplace=True)

data_uk = pd.concat([shadow_uk, inflation_uk, gdp_uk], axis=1).dropna()

print("\nMerged Data Columns (UK):")
print(data_uk.columns)

data_uk['INF UK'] = data_uk['INF UK'].pct_change() * 100
data_uk['GDP UK'] = data_uk['GDP UK'].pct_change() * 100

data_uk = data_uk.dropna()

print("\nAfter transformations (first few rows):")
print(data_uk.head())

def prepare_var_data(df):
    standardized_data = (df - df.mean()) / df.std()
    return standardized_data

var_data_uk = prepare_var_data(data_uk)

model_uk = VAR(var_data_uk)
lag_order = 2
results_uk = model_uk.fit(lag_order)

print("\nVAR Model Summary (UK):")
print(results_uk.summary())

n_periods = 20
irf_uk = results_uk.irf(n_periods)

print("\nImpulse Response Functions (UK system):")
fig = irf_uk.plot(orth=True)
for ax in fig.axes:
    ax.set_title(
        ax.get_title()
        .replace("INF UK", "CPI U.K.")
        .replace("GDP UK", "Real GDP U.K.")
        .replace("MAJD (United Kingdom)", "MAJD (U.K.)")
    )
plt.suptitle('Impulse Response Functions - MAJD (United Kingdom)')
plt.show()

def calculate_cumulative_irf(irf_obj, n_periods=10):
    cumulative_irf = {}
    irf_matrix = irf_obj.orth_irfs[:n_periods]
    response_names = irf_obj.model.names

    for i, impulse in enumerate(response_names):
        for j, response in enumerate(response_names):
            cumulative_irf[f"{response} → {impulse}"] = np.sum(np.abs(irf_matrix[:, j, i]))

    return cumulative_irf

cumulative_irf_uk = calculate_cumulative_irf(irf_uk)
cumulative_irf_uk_df = pd.DataFrame(cumulative_irf_uk, index=['Cumulative IRF'])

print("\nCumulative IRF Results (UK):")
print(cumulative_irf_uk_df.T)

cumulative_irf_uk_df.T.to_excel('cumulative_irf_final_shadow_uk.xlsx')
print("\nCumulative IRF saved to 'cumulative_irf_final_shadow_uk.xlsx'.")
```
