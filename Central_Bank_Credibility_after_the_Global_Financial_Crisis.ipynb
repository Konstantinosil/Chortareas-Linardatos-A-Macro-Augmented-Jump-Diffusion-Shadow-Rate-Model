```python
from __future__ import annotations

import importlib
import logging
import random
import subprocess
import sys
import warnings
from dataclasses import dataclass
from pathlib import Path
from typing import Dict, Optional, Sequence, Tuple, Union

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.tsa.api as smt
from scipy.stats import norm

try:
    from IPython.display import display
    from IPython import get_ipython
except Exception:
    get_ipython = None

    def display(obj):
        print(obj)


@dataclass(frozen=True)
class JournalConfig:
    ROOT_LOCAL: str = "."
    DATA_FILE: str = "Dataset for Credibility.xlsx"

    REPORTED_START: str = "2008-01-01"
    END: str = "2022-01-31"
    AP_PRESAMPLE_START: str = "2000-01-01"

    PI_STAR: float = 2.0
    ELB_EPS: float = 0.10
    ELB_THRESHOLDS: Optional[Dict[str, float]] = None

    RF_WINDOW_CAND: Tuple[int, ...] = (36, 48, 60, 72, 84)
    AP_SCALE_P_LO: int = 1
    AP_SCALE_P_HI: int = 99

    HAC_LAGS_BASE: int = 6
    WILD_BOOT_B: int = 399
    SEED: int = 123

    AR_MAX: int = 2
    MIN_T: int = 24
    FALLBACK_ROLL: int = 12
    ANCHOR_END: str = "2010-12-01"
    SHADOW_SCALE_P_LO: int = 5
    SHADOW_SCALE_P_HI: int = 95
    SHADOW_START: str = "2008-01-01"
    SHADOW_END: str = "2022-01-01"
    SHADOW_SEED: int = 42

    LEGACY_SHADOW_FORCE_AR0: bool = True
    LEGACY_SR_LAG_AFTER_DROP: bool = True

    HAMILTON_K: int = 12
    SR_H: int = 12
    SR_LAG: int = 12
    MIN_ABS_DPI: float = 0.10
    WINSOR_PCTS: Tuple[float, float] = (0.01, 0.99)

    PLOTS: bool = True
    SAVE_FIGURES: bool = False
    WRITE_RESULTS_TO_FILES: bool = False
    RESULTS_DIR: str = "credibility_results_single_workbook"

    AUTO_INSTALL_DEPENDENCIES: bool = True

    def __post_init__(self):
        if self.ELB_THRESHOLDS is None:
            object.__setattr__(
                self,
                "ELB_THRESHOLDS",
                {"ECB": 0.10, "BoE": 0.25, "Fed": 0.25},
            )


CFG = JournalConfig()

BANK_CODE_TO_NAME = {"EU": "ECB", "UK": "BoE", "US": "Fed"}
BANK_NAME_TO_CODE = {v: k for k, v in BANK_CODE_TO_NAME.items()}
BANK_ORDER = ["Fed", "ECB", "BoE"]
ECON_ORDER = ["US", "EU", "UK"]

def ensure_runtime_dependencies(auto_install: bool = True) -> None:
    missing = []
    for mod, pkg in [
        ("arch", "arch"),
        ("openpyxl", "openpyxl"),
        ("statsmodels", "statsmodels"),
    ]:
        if importlib.util.find_spec(mod) is None:
            missing.append(pkg)

    if missing:
        if not auto_install:
            raise ImportError(
                "Missing packages: "
                + ", ".join(missing)
                + ". Install them from the replication environment."
            )
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", *missing])


ensure_runtime_dependencies(CFG.AUTO_INSTALL_DEPENDENCIES)
from arch import arch_model


def activate_inline_backend() -> None:
    ip = get_ipython() if callable(get_ipython) else None
    if ip is not None:
        try:
            ip.run_line_magic("matplotlib", "inline")
        except Exception:
            pass


activate_inline_backend()

warnings.filterwarnings("ignore")
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings(
    "ignore",
    message="No frequency information was provided, so inferred frequency",
)
warnings.filterwarnings("ignore", category=RuntimeWarning)
logging.basicConfig(level=logging.INFO, format="[%(levelname)s] %(message)s")


def set_seeds(seed: int) -> None:
    np.random.seed(seed)
    random.seed(seed)


def resolve_input_file(filename: str, root_local: Optional[str] = None) -> Path:
    candidates = []
    if root_local:
        candidates.append(Path(root_local) / filename)
    candidates.extend(
        [
            Path.cwd() / filename,
            Path("/content") / filename,
            Path("/mnt/data") / filename,
            Path(filename),
        ]
    )

    seen = set()
    unique = []
    for p in candidates:
        key = str(p.resolve()) if p.exists() else str(p)
        if key not in seen:
            seen.add(key)
            unique.append(p)

    for p in unique:
        if p.exists():
            return p

    searched = "\n".join(str(x) for x in unique)
    raise FileNotFoundError(f"Could not find '{filename}'. Searched:\n{searched}")


def show_heading(title: str) -> None:
    bar = "=" * max(44, len(title))
    print(f"\n{bar}\n{title}\n{bar}")


def show_table(title: str, obj, max_rows: int = 200) -> None:
    show_heading(title)
    if obj is None:
        print("(none)")
        return
    if isinstance(obj, pd.Series):
        obj = obj.to_frame()
    if not hasattr(obj, "__len__") or len(obj) == 0:
        print("(empty)")
        return
    try:
        display(obj if len(obj) <= max_rows else obj.head(max_rows))
        if len(obj) > max_rows:
            print(f"Displayed first {max_rows} rows out of {len(obj)}.")
    except Exception:
        print(obj if len(obj) <= max_rows else obj.head(max_rows))


def month_end(ts) -> pd.Timestamp:
    return (
        pd.Timestamp(ts).to_period("M").to_timestamp(how="start")
        + pd.offsets.MonthEnd(0)
    ).normalize()


def month_start(ts) -> pd.Timestamp:
    return pd.Timestamp(ts).to_period("M").to_timestamp(how="start")


def save_fig(fig: plt.Figure, save_path: Optional[Path] = None, save_flag: bool = False):
    if save_flag and save_path is not None:
        save_path.parent.mkdir(parents=True, exist_ok=True)
        fig.savefig(save_path, bbox_inches="tight")
    plt.show()
    plt.close(fig)


def winsorize_df(df: pd.DataFrame, cols: Sequence[str], p: float = 0.01) -> pd.DataFrame:
    z = df.copy()
    for c in cols:
        if c in z.columns and z[c].notna().any():
            lo, hi = z[c].quantile([p, 1 - p])
            z[c] = z[c].clip(lo, hi)
    return z


def winsorize(s: pd.Series, p: Tuple[float, float] = (0.01, 0.99)) -> pd.Series:
    z = pd.Series(s).copy()
    if z.dropna().empty:
        return z
    lo, hi = z.quantile(p[0]), z.quantile(p[1])
    return z.clip(lo, hi)


def params_as_series(res) -> pd.Series:
    return pd.Series(res.params, index=res.model.exog_names).astype(float)


def bse_as_series(res) -> pd.Series:
    return pd.Series(res.bse, index=res.model.exog_names).astype(float)


def pvalues_as_series(res) -> pd.Series:
    return pd.Series(res.pvalues, index=res.model.exog_names).astype(float)


def safe_get_triplet(res, var: str) -> Tuple[float, float, float]:
    return (
        float(params_as_series(res).get(var, np.nan)),
        float(bse_as_series(res).get(var, np.nan)),
        float(pvalues_as_series(res).get(var, np.nan)),
    )


def merge_no_overlap(
    left: pd.DataFrame,
    right: pd.DataFrame,
    on: Sequence[str],
    suffix: str = "_r",
    how: str = "left",
    validate: Optional[str] = None,
) -> pd.DataFrame:
    on = list(on)
    overlap = (set(left.columns) & set(right.columns)) - set(on)
    r = right.rename(columns={c: f"{c}{suffix}" for c in overlap})
    return left.merge(r, on=on, how=how, validate=validate)


def scale_series_to_0_100(
    s: pd.Series,
    p_lo: int = 1,
    p_hi: int = 99,
) -> Tuple[pd.Series, Dict[str, float]]:
    x = pd.to_numeric(pd.Series(s), errors="coerce")
    base = x.dropna()
    if base.empty:
        out = pd.Series(np.nan, index=x.index, dtype=float)
        return out, {"q_lo": np.nan, "q_hi": np.nan}
    q_lo = float(np.nanpercentile(base.values, p_lo))
    q_hi = float(np.nanpercentile(base.values, p_hi))
    width = max(1e-8, q_hi - q_lo)
    scaled = ((x - q_lo) / width * 100.0).clip(0, 100)
    return scaled.astype(float), {"q_lo": q_lo, "q_hi": q_hi}

def parse_unified_sheet(name: str) -> Tuple[Optional[str], Optional[str]]:
    s = str(name).strip()
    parts = s.split()
    if len(parts) < 2:
        return None, None
    econ = parts[-1].upper()
    head = " ".join(parts[:-1]).lower()
    key = {
        "shadow rates": "SR",
        "policy rates": "POLICY",
        "inflation": "INF",
    }
    for k, v in key.items():
        if k in head:
            return v, econ
    if head.startswith("rgdp"):
        return "GDP", econ
    return None, None


def read_raw_sheet(path: Union[str, Path], sheet_name: str) -> pd.DataFrame:
    d = pd.read_excel(str(path), sheet_name=sheet_name)
    if d.shape[1] < 2:
        raise ValueError(f"Sheet '{sheet_name}' must have at least two columns.")

    d = d.iloc[:, :2].copy()
    d.columns = ["date", "value"]
    d["date"] = pd.to_datetime(d["date"], errors="coerce")

    raw = d["value"]
    num = pd.to_numeric(raw, errors="coerce")
    if num.notna().mean() < 0.5:
        num = (
            raw.astype(str)
            .str.replace("%", "", regex=False)
            .str.replace("\u2212", "-", regex=False)
            .str.replace(" ", "", regex=False)
            .str.replace(",", ".", regex=False)
        )
        num = pd.to_numeric(num, errors="coerce")

    d["value"] = num
    d = d.dropna(subset=["date"]).sort_values("date").reset_index(drop=True)
    return d


def load_unified_raw(path: Union[str, Path]) -> Tuple[Dict[Tuple[str, str], pd.DataFrame], pd.Series]:
    xls = pd.ExcelFile(str(path))
    raw: Dict[Tuple[str, str], pd.DataFrame] = {}
    t5y5y_daily = None

    for sh in xls.sheet_names:
        if sh.strip().lower() == "5y5y":
            fwd = pd.read_excel(str(path), sheet_name=sh)
            cols = {str(c).lower(): c for c in fwd.columns}
            date_col = cols.get("observation_date", fwd.columns[0])
            val_col = cols.get("t5yifr", fwd.columns[1])
            fwd = fwd[[date_col, val_col]].copy()
            fwd.columns = ["date", "t5y5y"]
            fwd["date"] = pd.to_datetime(fwd["date"], errors="coerce")
            fwd["t5y5y"] = pd.to_numeric(fwd["t5y5y"], errors="coerce")
            fwd = fwd.dropna(subset=["date", "t5y5y"]).sort_values("date")
            t5y5y_daily = fwd.set_index("date")["t5y5y"].rename("t5y5y")
            continue

        var, econ = parse_unified_sheet(sh)
        if var is None or econ not in BANK_CODE_TO_NAME:
            continue
        raw[(econ, var)] = read_raw_sheet(path, sh)

    required = [
        (econ, var)
        for econ in ["EU", "UK", "US"]
        for var in ["SR", "POLICY", "INF", "GDP"]
    ]
    missing = [k for k in required if k not in raw]
    if missing:
        raise KeyError(f"Missing required sheets/series: {missing}")
    if t5y5y_daily is None:
        raise KeyError("Missing required '5y5y' sheet with observation_date/T5YIFR.")
    return raw, t5y5y_daily


def freq_audit(xls_path: Union[str, Path]) -> pd.DataFrame:
    xls = pd.ExcelFile(str(xls_path))
    rows = []
    for sh in xls.sheet_names:
        d = pd.read_excel(xls, sheet_name=sh)
        if d.shape[1] < 2:
            rows.append(
                {
                    "sheet": sh,
                    "n_rows": len(d),
                    "n_dates": 0,
                    "n_values": 0,
                    "infer_freq": None,
                    "start": pd.NaT,
                    "end": pd.NaT,
                    "date_col": None,
                    "value_col": None,
                }
            )
            continue
        date_col = d.columns[0]
        val_col = d.columns[1]
        dd = pd.to_datetime(d[date_col], errors="coerce").dropna().sort_values()
        vv = pd.to_numeric(d[val_col], errors="coerce")
        try:
            freq = pd.infer_freq(dd[:20]) if len(dd) >= 10 else None
        except Exception:
            freq = None
        rows.append(
            {
                "sheet": sh,
                "n_rows": int(len(d)),
                "n_dates": int(dd.notna().sum()),
                "n_values": int(vv.notna().sum()),
                "infer_freq": freq,
                "start": dd.min(),
                "end": dd.max(),
                "date_col": str(date_col),
                "value_col": str(val_col),
            }
        )
    return pd.DataFrame(rows)


def to_monthly_eom(s: pd.Series, interpolate_quarterly: bool = True) -> pd.Series:
    z = pd.Series(s).copy()
    if not isinstance(z.index, pd.DatetimeIndex):
        z.index = pd.to_datetime(z.index, errors="coerce")
    z = z[~pd.isna(z.index)]
    z = pd.to_numeric(z, errors="coerce").sort_index()
    if z.index.has_duplicates:
        z = z.groupby(level=0).last()

    inferred = ""
    if len(z.dropna()) >= 3:
        try:
            inferred = (pd.infer_freq(z.dropna().index[:20]) or "").upper()
        except Exception:
            inferred = ""

    if inferred.startswith(("D", "B")):
        out = z.resample("ME").last()
        if out.index.has_duplicates:
            out = out.groupby(level=0).last()
        return out

    if inferred.startswith("Q"):
        q = z.copy()
        q.index = q.index.to_period("Q").to_timestamp("Q")
        q = q[~q.index.duplicated(keep="last")].sort_index()
        if not interpolate_quarterly:
            return q.resample("ME").last()
        monthly_idx = pd.date_range(
            q.index.min().to_period("M").to_timestamp("M"),
            q.index.max().to_period("M").to_timestamp("M"),
            freq="ME",
        )
        q_month = q.reindex(monthly_idx)
        q_month.loc[q.index] = q.values
        return q_month.interpolate(method="linear", limit_direction="both")

    z.index = z.index.to_period("M").to_timestamp("M")
    z = z.groupby(level=0).last().sort_index()
    return z.asfreq("ME")


def to_monthly_ms_legacy(s: pd.Series) -> pd.Series:
    z = pd.Series(s).copy()
    if not isinstance(z.index, pd.DatetimeIndex):
        z.index = pd.to_datetime(z.index, errors="coerce")
    z = z[~pd.isna(z.index)]
    z = pd.to_numeric(z, errors="coerce").sort_index()
    if z.index.has_duplicates:
        z = z.groupby(level=0).mean()
    try:
        f = (pd.infer_freq(z.dropna().index[:5]) or "").upper()
    except Exception:
        f = ""
    if f.startswith(("D", "B")):
        z = z.resample("MS").last()
    elif f.startswith("Q"):
        z = z.resample("QS").last().interpolate("linear").resample("MS").ffill()
    else:
        if not z.index.is_month_start.all():
            z.index = z.index.to_period("M").to_timestamp(how="start")
        z = z.asfreq("MS")
    return z


def ensure_ms_legacy(s: pd.Series) -> pd.Series:
    z = pd.Series(s).copy()
    if not isinstance(z.index, pd.DatetimeIndex):
        z.index = pd.to_datetime(z.index, errors="coerce")
    z = z[~pd.isna(z.index)]
    z.index = z.index.to_period("M").to_timestamp(how="start")
    z = pd.to_numeric(z, errors="coerce").sort_index()
    if z.index.has_duplicates:
        z = z.groupby(level=0).mean()
    return z.asfreq("MS")


def infer_yoy_from_INF(series: pd.Series) -> pd.Series:
    s = pd.to_numeric(pd.Series(series), errors="coerce").replace([np.inf, -np.inf], np.nan)
    med = s.dropna().median()

    if np.isfinite(med) and med > 30:
        s_pos = s.where(s > 0)
        return 100.0 * (np.log(s_pos) - np.log(s_pos.shift(12)))

    return s

def inflation_yoy_log_from_cpi(cpi: pd.Series) -> pd.Series:
    s = pd.to_numeric(pd.Series(cpi), errors="coerce")
    s = s.replace([np.inf, -np.inf], np.nan).dropna()
    l = np.log(s.clip(lower=1e-8))
    return 100.0 * (l - l.shift(12))


def inflation_yoy_from_cpi_legacy(cpi: pd.Series) -> pd.Series:
    cpi = ensure_ms_legacy(pd.Series(cpi).astype(float)).dropna()
    l = np.log(cpi.replace([np.inf, -np.inf], np.nan)).dropna()
    return 100.0 * (l - l.shift(12))


def hamilton_output_gap(log_gdp: pd.Series, k: int = 12) -> pd.Series:
    y = pd.to_numeric(log_gdp, errors="coerce").replace([np.inf, -np.inf], np.nan)
    y = y.sort_index()
    Xlags = pd.concat([y.shift(i) for i in range(1, k + 1)], axis=1)
    Xlags.columns = [f"lag{i}" for i in range(1, k + 1)]
    resid = pd.Series(index=y.index, dtype=float)
    for t in range(k + 8, len(y)):
        X_t = sm.add_constant(Xlags.iloc[:t].dropna(), has_constant="add")
        idx = X_t.index
        y_t = y.loc[idx]
        if len(y_t) >= k + 5:
            beta = sm.OLS(y_t, X_t).fit().params
            row = sm.add_constant(Xlags.iloc[t:t + 1], has_constant="add").iloc[0]
            y_hat = float((row * beta.reindex(row.index).fillna(0)).sum())
            resid.iloc[t] = y.iloc[t] - y_hat
    return resid * 100.0


def hamilton_output_gap_series_legacy(log_gdp: pd.Series, k: int = 12) -> pd.Series:
    y = pd.Series(log_gdp).copy()
    if not isinstance(y.index, pd.DatetimeIndex):
        y.index = pd.to_datetime(y.index, errors="coerce")
    y = pd.to_numeric(y, errors="coerce")
    y = y[~pd.isna(y.index)].sort_index()
    if y.index.has_duplicates:
        y = y.groupby(level=0).last()
    Xlags = pd.concat([y.shift(i) for i in range(1, k + 1)], axis=1)
    Xlags.columns = [f"lag{i}" for i in range(1, k + 1)]
    resid = pd.Series(index=y.index, dtype=float)
    for t in range(k + 8, len(y)):
        X_t = sm.add_constant(Xlags.iloc[:t].dropna(), has_constant="add")
        idx = X_t.index
        y_t = y.loc[idx]
        if len(y_t) >= k + 5:
            beta = sm.OLS(y_t, X_t).fit().params
            row = sm.add_constant(Xlags.iloc[t:t + 1], has_constant="add").iloc[0]
            y_hat = float((row * beta.reindex(row.index).fillna(0)).sum())
            resid.iloc[t] = y.iloc[t] - y_hat
    return (resid * 100.0).rename("gap")


def hamilton_gap_from_gdp_legacy(gdp: pd.Series, k: int = 12) -> pd.Series:
    s = pd.Series(gdp).copy()
    if not isinstance(s.index, pd.DatetimeIndex):
        s.index = pd.to_datetime(s.index, errors="coerce")
    s = pd.to_numeric(s, errors="coerce").replace([np.inf, -np.inf], np.nan)
    s = s[~pd.isna(s.index)].sort_index()
    if s.index.has_duplicates:
        s = s.groupby(level=0).last()
    s = s.ffill().bfill().clip(lower=1e-8)
    loglvl = np.log(s)
    return hamilton_output_gap_series_legacy(loglvl, k=k)


def qe_flag(bank: str, year: Union[int, float]) -> int:
    if pd.isna(year):
        return 0
    y = int(year)
    if bank == "ECB":
        return int((2015 <= y <= 2018) or (2020 <= y <= 2021))
    if bank == "BoE":
        return int((2009 <= y <= 2012) or (2020 <= y <= 2021))
    if bank == "Fed":
        return int((2009 <= y <= 2014) or (2020 <= y <= 2021))
    return 0


def build_common_macro_panel(
    raw: Dict[Tuple[str, str], pd.DataFrame],
    cfg: JournalConfig,
    elb_mode: str = "data_driven",
) -> pd.DataFrame:
    master_me = pd.date_range(month_end(cfg.AP_PRESAMPLE_START), month_end(cfg.END), freq="ME")
    rows = []
    for econ in ["EU", "UK", "US"]:
        bank = BANK_CODE_TO_NAME[econ]
        pol = raw[(econ, "POLICY")]
        inf = raw[(econ, "INF")]
        gdp = raw[(econ, "GDP")]

        m_pol = to_monthly_eom(pol.set_index("date")["value"]).reindex(master_me)
        inf_level = to_monthly_eom(inf.set_index("date")["value"], interpolate_quarterly=True).reindex(master_me)
        infl_yoy = infer_yoy_from_INF(inf_level)
        pi_log_yoy = infl_yoy.copy()
        m_gdp = to_monthly_eom(gdp.set_index("date")["value"], interpolate_quarterly=True).reindex(master_me)
        tmp = pd.concat([m_pol, inf_level, infl_yoy, pi_log_yoy, m_gdp], axis=1)
        tmp.columns = ["policy_rate", "inf_level", "infl_yoy", "PI_log", "gdp"]
        tmp["bank"] = bank
        tmp["econ"] = econ
        tmp = tmp.reset_index().rename(columns={"index": "date"})
        rows.append(tmp)

    mon = pd.concat(rows, ignore_index=True).sort_values(["bank", "date"]).reset_index(drop=True)

    outs = []
    for bank, g in mon.groupby("bank", sort=False):
        gg = g.copy()
        gg["gdp"] = pd.to_numeric(gg["gdp"], errors="coerce").ffill().bfill()
        gg["log_gdp"] = np.log(gg["gdp"].clip(lower=1e-8))
        gap = hamilton_output_gap(gg.set_index("date")["log_gdp"], k=cfg.HAMILTON_K)
        gg["output_gap"] = gap.reindex(gg["date"]).values
        outs.append(gg)
    mon = pd.concat(outs, ignore_index=True).sort_values(["bank", "date"]).reset_index(drop=True)

    mon["inflation_gap"] = pd.to_numeric(mon["infl_yoy"], errors="coerce") - cfg.PI_STAR
    mon["year"] = mon["date"].dt.year
    mon["month"] = mon["date"].dt.month
    mon["ym"] = mon["date"].dt.to_period("M")
    mon["crisis10"] = mon["year"].between(2010, 2012).astype(int)
    mon["pandem"] = mon["year"].between(2020, 2021).astype(int)
    mon["QE_FG"] = mon.apply(lambda r: qe_flag(r["bank"], r["year"]), axis=1).astype(int)

    if elb_mode == "data_driven":
        def _elb_transform(s: pd.Series) -> pd.Series:
            x = pd.to_numeric(s, errors="coerce")
            roll_min = x.rolling(24, min_periods=24).min()
            return (x <= (roll_min + cfg.ELB_EPS)).astype(int)

        mon["ELB"] = (
            mon.groupby("bank", group_keys=False)["policy_rate"]
            .transform(_elb_transform)
            .astype(int)
        )
    else:
        mon["ELB"] = mon.apply(
            lambda r: int(r["policy_rate"] <= cfg.ELB_THRESHOLDS.get(str(r["bank"]), 0.25)),
            axis=1,
        ).astype(int)

    mon["policy_rate_lead3"] = mon.groupby("bank")["policy_rate"].shift(-3)
    mon["di"] = mon.groupby("bank")["policy_rate"].diff()
    mon["abs_deltai_3m"] = (mon["policy_rate_lead3"] - mon["policy_rate"]).abs()
    mon["abs_deltai_3m_lag"] = mon.groupby("bank")["abs_deltai_3m"].shift(1)
    mon["abs_pi_gap"] = mon["inflation_gap"].abs()
    mon["abs_y_gap"] = mon["output_gap"].abs()
    den = mon["abs_pi_gap"] + mon["abs_y_gap"]
    mon["pi_share"] = np.where(den > 0, mon["abs_pi_gap"] / den, np.nan)
    return mon


def macro_backbone_audit(panel: pd.DataFrame) -> pd.DataFrame:
    rows = []
    for bank, g in panel.groupby("bank"):
        for c in [
            "policy_rate",
            "inf_level",
            "infl_yoy",
            "PI_log",
            "gdp",
            "output_gap",
            "inflation_gap",
            "ELB",
            "pi_share",
        ]:
            s = g[c] if c in g.columns else pd.Series(dtype=float)
            rows.append(
                {
                    "bank": bank,
                    "series": c,
                    "nonmissing": int(s.notna().sum()),
                    "start": g.loc[s.notna(), "date"].min() if s.notna().any() else pd.NaT,
                    "end": g.loc[s.notna(), "date"].max() if s.notna().any() else pd.NaT,
                    "min": float(pd.to_numeric(s, errors="coerce").min()) if s.notna().any() else np.nan,
                    "max": float(pd.to_numeric(s, errors="coerce").max()) if s.notna().any() else np.nan,
                }
            )
    return pd.DataFrame(rows)


def estimate_reaction_function_realtime(mon: pd.DataFrame, window: int = 60) -> pd.DataFrame:
    Z = ["inflation_gap", "output_gap", "ELB", "QE_FG", "crisis10", "pandem", "di_lag"]
    out = []
    for bank, g in mon.sort_values(["bank", "date"]).groupby("bank"):
        z = g.copy()
        z["di_lag"] = z["di"].shift(1)
        z = z.dropna(subset=["di"] + Z)
        if z.empty:
            continue
        names = ["const"] + Z
        di_hat = []
        for i in range(len(z)):
            lo = max(0, i - window)
            sub = z.iloc[lo:i]
            if len(sub) < max(24, len(Z) + 5):
                di_hat.append(np.nan)
                continue
            X = pd.DataFrame({"const": 1.0}, index=sub.index)
            for c in Z:
                X[c] = pd.to_numeric(sub[c], errors="coerce").astype(float)
            res = sm.OLS(pd.to_numeric(sub["di"], errors="coerce").astype(float), X).fit()
            x_now = pd.Series({"const": 1.0}, dtype=float)
            for c in Z:
                x_now[c] = float(z.iloc[i][c])
            di_hat.append(float((x_now[names] * res.params.reindex(names).fillna(0)).sum()))
        z["di_hat_rt"] = di_hat
        z["di_surpr_rt"] = z["di"] - z["di_hat_rt"]
        out.append(z[["date", "bank", "di_hat_rt", "di_surpr_rt"]])
    if not out:
        return pd.DataFrame(columns=["date", "bank", "di_hat_rt", "di_surpr_rt"])
    return pd.concat(out, ignore_index=True)


def build_surprise_h_rt(mon: pd.DataFrame, h: int = 3) -> pd.DataFrame:
    if h != 3:
        raise ValueError("Original First Credibility Metric is centered AP(3). Use h=3.")
    z = mon.sort_values(["bank", "date"]).copy()
    u_lag1 = z.groupby("bank")["di_surpr_rt"].shift(1)
    u_now = z.groupby("bank")["di_surpr_rt"].shift(0)
    u_lead1 = z.groupby("bank")["di_surpr_rt"].shift(-1)
    z["cum_surpr_3_rt"] = pd.concat([u_lag1, u_now, u_lead1], axis=1).sum(axis=1, min_count=3)
    z["abs_surprise_rel_3_rt"] = z["cum_surpr_3_rt"].abs()
    return z


def rf_params_rt(mon: pd.DataFrame, window: int = 60) -> pd.DataFrame:
    Z = ["inflation_gap", "output_gap", "ELB", "QE_FG", "crisis10", "pandem", "di_lag"]
    out = []
    for bank, g in mon.sort_values(["bank", "date"]).groupby("bank"):
        z = g.copy()
        z["di_lag"] = z["di"].shift(1)
        z = z.dropna(subset=["di"] + Z)
        phi_pi, phi_y = [], []
        for i in range(len(z)):
            lo = max(0, i - window)
            sub = z.iloc[lo:i]
            if len(sub) < max(24, len(Z) + 5):
                phi_pi.append(np.nan)
                phi_y.append(np.nan)
                continue
            X = pd.DataFrame({"const": 1.0}, index=sub.index)
            for c in Z:
                X[c] = pd.to_numeric(sub[c], errors="coerce").astype(float)
            params = sm.OLS(pd.to_numeric(sub["di"], errors="coerce").astype(float), X).fit().params
            phi_pi.append(float(params.get("inflation_gap", np.nan)))
            phi_y.append(float(params.get("output_gap", np.nan)))
        tmp = z[["date"]].copy()
        tmp["bank"] = bank
        tmp["phi_pi_rt"] = phi_pi
        tmp["phi_y_rt"] = phi_y
        den = np.abs(tmp["phi_pi_rt"]) + np.abs(tmp["phi_y_rt"])
        tmp["omega_t_rt"] = np.where(den > 0, np.abs(tmp["phi_pi_rt"]) / den, np.nan)
        out.append(tmp)
    return pd.concat(out, ignore_index=True) if out else pd.DataFrame()


def cv_select_rf_window_rt(mon: pd.DataFrame, cand: Sequence[int]) -> pd.DataFrame:
    rows = []
    for w in cand:
        rf = estimate_reaction_function_realtime(mon, window=w)
        z = mon.merge(rf, on=["bank", "date"], how="left")
        mse = np.nanmean(z["di_surpr_rt"] ** 2)
        rows.append(
            {
                "window": int(w),
                "mse": float(mse) if np.isfinite(mse) else np.inf,
                "n_surprise": int(z["di_surpr_rt"].notna().sum()),
            }
        )
    return pd.DataFrame(rows).sort_values(["mse", "n_surprise"], ascending=[True, False]).reset_index(drop=True)


def plot_ap3_credibility_score(z_reported: pd.DataFrame, cfg: JournalConfig) -> None:
    if not cfg.PLOTS:
        return
    show_heading("Figure 1A. First Credibility Score")
    fig, axes = plt.subplots(3, 1, figsize=(11, 9), sharex=True)
    for ax, bank in zip(axes, BANK_ORDER):
        g = z_reported[z_reported["bank"] == bank].dropna(subset=["cred_ap3_score"]).copy()
        if g.empty:
            ax.set_visible(False)
            continue
        ax.plot(g["date"], g["cred_ap3_score"])
        ax.set_title(f"Rule-Relative Adjustment Pressure — {bank}")
        ax.set_ylabel("Credibility")
        ax.set_xlim(month_end(cfg.REPORTED_START), month_end(cfg.END))
        ax.set_ylim(0, 100)
        ax.grid(True, linestyle="--", linewidth=0.5, alpha=0.4)
    axes[-1].set_xlabel("Date")
    fig.tight_layout()
    save_fig(
        fig,
        Path(cfg.RESULTS_DIR) / "figure_first_credibility_score.png",
        cfg.SAVE_FIGURES,
    )


def build_ap3_measure(common_panel: pd.DataFrame, cfg: JournalConfig) -> Dict[str, object]:
    backbone = common_panel.copy()
    cv_table = cv_select_rf_window_rt(backbone, cfg.RF_WINDOW_CAND)
    best_w = int(cv_table.iloc[0]["window"])
    rf = estimate_reaction_function_realtime(backbone, window=best_w)
    mon = merge_no_overlap(backbone, rf, on=["bank", "date"], suffix="_rf", validate="one_to_one")
    if "di_surpr_rt_rf" in mon.columns:
        mon["di_surpr_rt"] = mon["di_surpr_rt_rf"]
    if "di_hat_rt_rf" in mon.columns:
        mon["di_hat_rt"] = mon["di_hat_rt_rf"]
    mon = build_surprise_h_rt(mon, h=3)
    mon["AP3_raw_lowcred"] = mon["abs_surprise_rel_3_rt"]
    mon["lowcred_ap3"] = mon["AP3_raw_lowcred"]
    phidf = rf_params_rt(mon, window=best_w)
    if not phidf.empty:
        mon = mon.merge(phidf, on=["bank", "date"], how="left", validate="one_to_one")
    mon_reported = mon.loc[mon["date"] >= month_end(cfg.REPORTED_START)].copy()
    mon_reported["AP3_lowcred_scaled"], ap3_scale_stats = scale_series_to_0_100(
        mon_reported["AP3_raw_lowcred"],
        p_lo=cfg.AP_SCALE_P_LO,
        p_hi=cfg.AP_SCALE_P_HI,
    )
    mon_reported["cred_ap3_score"] = 100.0 - mon_reported["AP3_lowcred_scaled"]
    mon = mon.merge(
        mon_reported[["date", "bank", "AP3_lowcred_scaled", "cred_ap3_score"]],
        on=["date", "bank"],
        how="left",
        validate="one_to_one",
    )
    plot_ap3_credibility_score(mon_reported, cfg)
    return {
        "panel_full": mon,
        "panel_reported": mon_reported,
        "rf_window_cv": cv_table,
        "best_rf_window": best_w,
        "scale_stats": ap3_scale_stats,
    }


def build_shadow_rate_wide(raw: Dict[Tuple[str, str], pd.DataFrame], cfg: JournalConfig) -> pd.DataFrame:
    idx = pd.date_range(month_start(cfg.SHADOW_START), month_start(cfg.SHADOW_END), freq="MS")
    out = {}
    for econ in ["US", "EU", "UK"]:
        sr = raw[(econ, "SR")]
        out[econ] = to_monthly_ms_legacy(sr.set_index("date")["value"]).reindex(idx)
    return pd.DataFrame(out).sort_index()


def select_ar_shadow(sr: pd.Series, cfg: JournalConfig) -> int:
    if cfg.LEGACY_SHADOW_FORCE_AR0:
        return 0
    z = ensure_ms_legacy(pd.Series(sr).dropna())
    best, lag, success = np.inf, 0, 0
    for p in range(cfg.AR_MAX + 1):
        try:
            fit = smt.ARIMA(z, order=(p, 0, 0)).fit(method="innovations_mle")
            success += 1
            if fit.aic < best:
                best, lag = float(fit.aic), int(p)
        except Exception:
            pass
    if success == 0:
        logging.warning("Repaired AIC selection failed for all p; using p=0 fallback.")
    return int(lag)


def fallback_vol_ms(y: pd.Series, roll: int) -> pd.Series:
    z = ensure_ms_legacy(pd.Series(y).astype(float))
    return z.rolling(roll).std(ddof=0)


def fit_egarch_t_level(
    y: pd.Series,
    lags_ar: int,
    min_T: int,
    fallback_roll: int,
):
    z = ensure_ms_legacy(pd.Series(y).astype(float).dropna())
    if len(z) < min_T:
        return None, fallback_vol_ms(z, fallback_roll)
    mu, sd = z.mean(), z.std(ddof=0)
    if (not np.isfinite(sd)) or sd == 0:
        return None, fallback_vol_ms(z, fallback_roll)
    y_std = (z - mu) / sd
    try:
        am = arch_model(
            y_std,
            mean="AR",
            lags=lags_ar,
            vol="EGARCH",
            p=1,
            o=0,
            q=1,
            dist="t",
            rescale=True,
        )
        res = am.fit(disp="off", show_warning=False, options={"maxiter": 2500})
        sig = pd.Series(res.conditional_volatility, index=z.index).replace([np.inf, -np.inf], np.nan)
        if sig.isna().mean() > 0.20:
            return None, fallback_vol_ms(z, fallback_roll)
        return res, sig * sd
    except Exception as e:
        logging.warning(f"EGARCH failed: {e}; fallback rolling std.")
        return None, fallback_vol_ms(z, fallback_roll)


def fit_gjr_garch_t_level(
    y: pd.Series,
    lags_ar: int,
    min_T: int,
    fallback_roll: int,
):
    z = ensure_ms_legacy(pd.Series(y).astype(float).dropna())
    if len(z) < min_T:
        return None, fallback_vol_ms(z, fallback_roll)
    mu, sd = z.mean(), z.std(ddof=0)
    if (not np.isfinite(sd)) or sd == 0:
        return None, fallback_vol_ms(z, fallback_roll)
    y_std = (z - mu) / sd
    try:
        am = arch_model(
            y_std,
            mean="AR",
            lags=lags_ar,
            vol="GARCH",
            p=1,
            o=1,
            q=1,
            dist="t",
            rescale=True,
        )
        res = am.fit(disp="off", show_warning=False, options={"maxiter": 2500})
        sig = pd.Series(res.conditional_volatility, index=z.index).replace([np.inf, -np.inf], np.nan)
        if sig.isna().mean() > 0.20:
            return None, fallback_vol_ms(z, fallback_roll)
        return res, sig * sd
    except Exception as e:
        logging.warning(f"GJR-GARCH failed: {e}; fallback rolling std.")
        return None, fallback_vol_ms(z, fallback_roll)


def scale_shadow_wide_to_0_100_legacy(
    rawcred: pd.DataFrame,
    anchor_end: str,
    p_lo: int = 5,
    p_hi: int = 95,
) -> Tuple[pd.DataFrame, Dict[str, float]]:
    anchor_end_ts = pd.Timestamp(anchor_end)
    z = rawcred.copy()
    z.index = pd.to_datetime(z.index).to_period("M").to_timestamp(how="start")
    base = z.loc[z.index <= anchor_end_ts].stack().dropna()
    if base.empty:
        raise ValueError("Anchor window empty. Check ANCHOR_END.")
    q_lo = float(np.nanpercentile(base.values, p_lo))
    q_hi = float(np.nanpercentile(base.values, p_hi))
    width = max(1e-8, q_hi - q_lo)
    scaled = ((z - q_lo) / width * 100.0).clip(0, 100)
    stats = {"q_lo": q_lo, "q_hi": q_hi, "anchor_end": str(anchor_end_ts.date())}
    return scaled, stats


def plot_shadow_vol_and_cred(
    sigma_df: pd.DataFrame,
    cred_df: pd.DataFrame,
    cfg: JournalConfig,
    econ_subset: Optional[Sequence[str]] = None,
) -> None:
    if not cfg.PLOTS:
        return
    label_map = {"US": "Fed", "EU": "ECB", "UK": "BoE"}
    if econ_subset is None:
        econ_subset = ["US", "EU", "UK"]
    econ_subset = [e.upper() for e in econ_subset]
    t0 = month_start(cfg.SHADOW_START)
    t1 = month_start(cfg.SHADOW_END)

    fig, axes = plt.subplots(len(econ_subset), 1, figsize=(11, 9), sharex=True)
    if len(econ_subset) == 1:
        axes = [axes]
    for ax, econ in zip(axes, econ_subset):
        s = sigma_df[econ].dropna()
        s = s.loc[s.index >= t0]
        if s.empty:
            ax.set_visible(False)
            continue
        ax.plot(s.index, s.values)
        ax.set_title(f"EGARCH-t shadow-rate volatility — {label_map.get(econ, econ)}")
        ax.set_ylabel("Volatility")
        ax.set_xlim(t0, t1)
        ax.grid(True, linestyle="--", linewidth=0.5, alpha=0.4)
    axes[-1].set_xlabel("Date")
    fig.tight_layout()
    save_fig(
        fig,
        Path(cfg.RESULTS_DIR) / "figure_shadow_volatility.png",
        cfg.SAVE_FIGURES,
    )

    fig, axes = plt.subplots(len(econ_subset), 1, figsize=(11, 9), sharex=True)
    if len(econ_subset) == 1:
        axes = [axes]
    for ax, econ in zip(axes, econ_subset):
        s = cred_df[econ].dropna()
        s = s.loc[s.index >= t0]
        if s.empty:
            ax.set_visible(False)
            continue
        ax.plot(s.index, s.values)
        ax.set_title(f"Shadow-Rate Stance Stability — {label_map.get(econ, econ)}")
        ax.set_ylabel("Credibility")
        ax.set_ylim(0, 100)
        ax.set_xlim(t0, t1)
        ax.grid(True, linestyle="--", linewidth=0.5, alpha=0.4)
    axes[-1].set_xlabel("Date")
    fig.tight_layout()
    save_fig(
        fig,
        Path(cfg.RESULTS_DIR) / "figure_second_credibility_score.png",
        cfg.SAVE_FIGURES,
    )


def run_shadow_rate_measure(
    shadow_wide: pd.DataFrame,
    cfg: JournalConfig,
) -> Dict[str, object]:
    set_seeds(cfg.SHADOW_SEED)
    show_heading("WU & XIA SHADOW-RATE PIPELINE — LEGACY 1:1 SINGLE WORKBOOK")

    ar_lags: Dict[str, int] = {}
    sigma_eg: Dict[str, pd.Series] = {}
    eg_meta: Dict[str, Dict[str, object]] = {}
    for e in ["US", "EU", "UK"]:
        sr = ensure_ms_legacy(shadow_wide[e]).dropna()
        p = select_ar_shadow(sr, cfg)
        ar_lags[e] = int(p)
        res, sig = fit_egarch_t_level(
            sr,
            lags_ar=p,
            min_T=cfg.MIN_T,
            fallback_roll=cfg.FALLBACK_ROLL,
        )
        sigma_eg[e] = sig
        eg_meta[e] = {
            "ar_lags": int(p),
            "ar_source": "legacy_forced_AR0" if cfg.LEGACY_SHADOW_FORCE_AR0 else "AIC_repaired",
            "egarch_ok": bool(res is not None),
        }

    sigma_eg_df = pd.DataFrame(sigma_eg).sort_index()
    cred_eg_raw = (-np.log(sigma_eg_df.clip(lower=1e-10))).rename_axis("date")
    cred_eg_sc, anchor_stats = scale_shadow_wide_to_0_100_legacy(
        cred_eg_raw,
        cfg.ANCHOR_END,
        cfg.SHADOW_SCALE_P_LO,
        cfg.SHADOW_SCALE_P_HI,
    )

    print("\n=== EGARCH meta (baseline SR-level, legacy 1:1) ===")
    print(pd.DataFrame(eg_meta).T)
    print("\n=== Anchor stats (pooled scaling, legacy WX) ===")
    print(anchor_stats)
    if cfg.LEGACY_SHADOW_FORCE_AR0:
        print("\nLegacy note: AR lag is fixed at p=0 to reproduce the original code's effective behavior.")

    sigma_gjr: Dict[str, pd.Series] = {}
    gjr_meta: Dict[str, Dict[str, object]] = {}
    for e in ["US", "EU", "UK"]:
        sr = ensure_ms_legacy(shadow_wide[e]).dropna()
        p = ar_lags[e]
        res, sig = fit_gjr_garch_t_level(
            sr,
            lags_ar=p,
            min_T=cfg.MIN_T,
            fallback_roll=cfg.FALLBACK_ROLL,
        )
        sigma_gjr[e] = sig
        gjr_meta[e] = {"ar_lags": int(p), "gjr_ok": bool(res is not None)}

    sigma_gjr_df = pd.DataFrame(sigma_gjr).sort_index()
    cred_gjr_raw = (-np.log(sigma_gjr_df.clip(lower=1e-10))).rename_axis("date")
    cred_gjr_sc, _ = scale_shadow_wide_to_0_100_legacy(
        cred_gjr_raw,
        cfg.ANCHOR_END,
        cfg.SHADOW_SCALE_P_LO,
        cfg.SHADOW_SCALE_P_HI,
    )

    rows = []
    for e in ["US", "EU", "UK"]:
        tmp = pd.concat(
            [cred_eg_sc[e].rename("egarch"), cred_gjr_sc[e].rename("gjr")],
            axis=1,
        ).dropna()
        rows.append(
            [
                e,
                float(tmp["egarch"].corr(tmp["gjr"])) if len(tmp) > 5 else np.nan,
                int(len(tmp)),
            ]
        )
    pooled = pd.concat(
        [cred_eg_sc.stack().rename("egarch"), cred_gjr_sc.stack().rename("gjr")],
        axis=1,
    ).dropna()
    print("\n=== Robustness: baseline EGARCH credibility vs GJR-GARCH credibility ===")
    print(pd.DataFrame(rows, columns=["econ", "corr(egarch,gjr)", "T_common"]).to_string(index=False))
    print(f"Pooled corr(egarch,gjr) = {float(pooled['egarch'].corr(pooled['gjr'])):.3f} (N={len(pooled)})")
    print("\n=== GJR meta ===")
    print(pd.DataFrame(gjr_meta).T)

    plot_shadow_vol_and_cred(sigma_eg_df, cred_eg_sc, cfg)

    return {
        "sigma_eg": sigma_eg_df,
        "cred_eg_raw": cred_eg_raw,
        "cred_eg_sc": cred_eg_sc,
        "eg_meta": pd.DataFrame(eg_meta).T,
        "anchor_stats": anchor_stats,
        "sigma_gjr": sigma_gjr_df,
        "cred_gjr_raw": cred_gjr_raw,
        "cred_gjr_sc": cred_gjr_sc,
        "gjr_meta": pd.DataFrame(gjr_meta).T,
        "gjr_corr_by_econ": pd.DataFrame(rows, columns=["econ", "corr(egarch,gjr)", "T_common"]),
        "gjr_corr_pooled": float(pooled["egarch"].corr(pooled["gjr"])) if len(pooled) > 0 else np.nan,
    }

def legacy_load_panel_from_raw(
    raw: Dict[Tuple[str, str], pd.DataFrame],
    econ_list: Sequence[str],
    start: str,
    end: str,
) -> pd.DataFrame:
    panel_raw = {}
    for econ in econ_list:
        for var in ["SR", "INF", "GDP"]:
            d = raw[(econ, var)]
            s = to_monthly_ms_legacy(d.set_index("date")["value"])
            panel_raw[(econ, var)] = s.loc[str(start):str(end)]
    panel = (
        pd.concat(panel_raw, axis=1)
        .swaplevel(0, 1, axis=1)
        .sort_index(axis=1)
        .stack(level=1)
        .rename_axis(["date", "econ"])
        .sort_index()
    )
    if panel.index.duplicated().any():
        panel = panel.groupby(level=["date", "econ"]).mean()
    return panel


def prep_shadow_panel_legacy(
    raw: Dict[Tuple[str, str], pd.DataFrame],
    cfg: JournalConfig,
) -> pd.DataFrame:
    panel = legacy_load_panel_from_raw(
        raw,
        econ_list=["US", "EU", "UK"],
        start=cfg.SHADOW_START,
        end=cfg.SHADOW_END,
    )
    needed = {"SR", "INF", "GDP"}
    missing = needed - set(panel.columns)
    if missing:
        raise KeyError(f"Missing variables in legacy WX panel: {missing}")

    print("=== Legacy WX data coverage (non-missing counts) ===")
    for v in ["SR", "INF", "GDP"]:
        cnt = panel[v].groupby("econ").apply(lambda s: int(s.notna().sum()))
        print(v, dict(cnt))

    inf_wide = panel["INF"].unstack("econ").sort_index()
    inf_wide = inf_wide.loc[~inf_wide.index.duplicated(keep="last")]
    pi_wide = inf_wide.apply(inflation_yoy_from_cpi_legacy, axis=0)
    pi_long = pi_wide.stack().rename("PI").groupby(level=["date", "econ"]).mean()

    panel = (
        panel.reset_index()
        .merge(pi_long.to_frame("PI").reset_index(), on=["date", "econ"], how="left")
        .set_index(["date", "econ"])
        .sort_index()
    )
    if panel.index.duplicated().any():
        panel = panel.groupby(level=["date", "econ"]).mean()
    return panel


def sacrifice_ratio_legacy(
    pi: pd.Series,
    gap: pd.Series,
    H: int,
    min_abs_dpi: float,
    winsor_pcts: Tuple[float, float],
) -> pd.Series:
    pi = ensure_ms_legacy(pi).astype(float).sort_index()
    gap = ensure_ms_legacy(gap).astype(float).sort_index()
    cost_pm = gap.clip(upper=0).abs().rolling(H, min_periods=H).sum()
    cost_py = cost_pm / 12.0
    dpi = pi - pi.shift(H)
    sr = (cost_py / dpi.abs()).where(dpi < -min_abs_dpi)
    return winsorize(sr, winsor_pcts).rename("SR")


def run_sr_reg_hac_only(
    sr_panel_long: pd.DataFrame,
    cred_col: str,
    H: int,
    L: int,
    analysis_start: Optional[str] = None,
):
    df = sr_panel_long.copy().sort_values(["econ", "date"])
    df[f"cred_lag{L}"] = df.groupby("econ")[cred_col].shift(L)
    if analysis_start is not None:
        df = df[df["date"] >= pd.Timestamp(analysis_start)].copy()
    use = df[["SR", f"cred_lag{L}", "econ"]].dropna().copy()
    D = pd.get_dummies(use["econ"], prefix="ent", drop_first=True, dtype=int)
    X = pd.concat([use[[f"cred_lag{L}"]], D], axis=1)
    X = sm.add_constant(X, has_constant="add")
    y = use["SR"]
    res_hac = sm.OLS(y, X).fit(cov_type="HAC", cov_kwds={"maxlags": H})
    return res_hac, int(res_hac.nobs), use


def coef_line(res, var: str) -> Tuple[float, float, float]:
    b = float(res.params.get(var, np.nan))
    se = float(res.bse.get(var, np.nan))
    p = float(res.pvalues.get(var, np.nan))
    return b, se, p


def run_sacrifice_ratio_validation_legacy(
    shadow_wx_panel: pd.DataFrame,
    cred_eg_raw: pd.DataFrame,
    cfg: JournalConfig,
) -> Dict[str, object]:
    sr_rows = []
    for e in ["US", "EU", "UK"]:
        sub = shadow_wx_panel.xs(e, level="econ").sort_index()
        gap = hamilton_gap_from_gdp_legacy(sub["GDP"], k=cfg.HAMILTON_K)
        pi = sub["PI"]
        srH = sacrifice_ratio_legacy(
            pi,
            gap,
            H=cfg.SR_H,
            min_abs_dpi=cfg.MIN_ABS_DPI,
            winsor_pcts=cfg.WINSOR_PCTS,
        )
        sr_rows.append(pd.DataFrame({"date": srH.index, "econ": e, "SR": srH.values}))

    sr_df = pd.concat(sr_rows, axis=0)
    if cfg.LEGACY_SR_LAG_AFTER_DROP:
        sr_df = sr_df.dropna(subset=["SR"])
    sr_df = sr_df.sort_values(["econ", "date"]).reset_index(drop=True)

    cred_long = cred_eg_raw.stack().rename("cred_eg_raw")
    cred_long.index.names = ["date", "econ"]
    sr_df = (
        sr_df.merge(cred_long.reset_index(), on=["date", "econ"], how="left")
        .dropna(subset=["cred_eg_raw"])
        .sort_values(["econ", "date"])
        .reset_index(drop=True)
    )

    H = cfg.SR_H
    L = cfg.SR_LAG
    xname = f"cred_lag{L}"
    res_hac_eg, N_eg, use = run_sr_reg_hac_only(
        sr_df,
        cred_col="cred_eg_raw",
        H=H,
        L=L,
        analysis_start=cfg.SHADOW_START,
    )

    print("\n==================== SACRIFICE RATIO REGRESSION (WX only, LEGACY 1:1) ====================")
    print(f"Model: SR_H{H} ~ const + cred(t-{L}) + country dummies (pooled)")
    print("Notes: SR uses overlapping windows; HAC(maxlags=H) is conservative under persistence.")
    if cfg.LEGACY_SR_LAG_AFTER_DROP:
        print("Legacy timing: SR non-missing windows are selected before cred_lag is computed.\n")
    else:
        print("Corrected timing: cred_lag is computed on the full monthly panel before SR windows are dropped.\n")

    b, se, p = coef_line(res_hac_eg, xname)
    print(f"[Baseline cred = EGARCH] HAC: b={b:.4f}, se={se:.4f}, p={p:.4f} | N={N_eg} | maxlags={H}")
    print("\n--- EGARCH credibility regression table (HAC) ---")
    print(res_hac_eg.summary().tables[1])

    return {
        "sr_df": sr_df,
        "sr_reg_use": use,
        "res_hac_eg": res_hac_eg,
        "N_eg": N_eg,
    }

def fe_ols(
    y: str,
    X: Sequence[str],
    data: pd.DataFrame,
    bank_fe: bool = True,
    year_fe: bool = False,
    month_fe: bool = False,
    bank_year_fe: bool = False,
    hac_lags: Optional[int] = None,
):
    if bank_year_fe:
        bank_fe = False
    z = data.copy()
    X_df = pd.DataFrame(index=z.index)
    X_df["const"] = 1.0
    for name in X:
        X_df[name] = pd.to_numeric(z[name], errors="coerce")
    if bank_fe and "bank" in z.columns:
        d = pd.get_dummies(z["bank"], prefix="", prefix_sep="", drop_first=True).astype(float)
        X_df = pd.concat([X_df, d], axis=1)
    if bank_year_fe and ("bank" in z.columns) and ("year" in z.columns):
        by = z["bank"].astype(str) + "_" + z["year"].astype(str)
        d = pd.get_dummies(by, prefix="by", drop_first=True).astype(float)
        X_df = pd.concat([X_df, d], axis=1)
    else:
        if year_fe and "year" in z.columns:
            d = pd.get_dummies(z["year"], prefix="y", drop_first=True).astype(float)
            X_df = pd.concat([X_df, d], axis=1)
    if month_fe and "month" in z.columns:
        d = pd.get_dummies(z["month"], prefix="m", drop_first=True).astype(float)
        X_df = pd.concat([X_df, d], axis=1)

    X_df = X_df.apply(pd.to_numeric, errors="coerce").replace([np.inf, -np.inf], np.nan)
    X_df = X_df.loc[:, X_df.notna().any(axis=0)]
    yv = pd.to_numeric(z[y], errors="coerce").replace([np.inf, -np.inf], np.nan)
    mask = yv.notna() & X_df.notna().all(axis=1)
    X_df = X_df.loc[mask].astype(float)
    yv = yv.loc[mask].astype(float)
    if X_df.empty or yv.empty:
        raise ValueError("Empty design after NA drop.")
    res = sm.OLS(yv, X_df).fit(
        cov_type="HAC",
        cov_kwds={"maxlags": hac_lags or CFG.HAC_LAGS_BASE},
    )
    res._index = X_df.index.copy()
    return res


def _cov_as_df(res) -> pd.DataFrame:
    cov = res.cov_params()
    if isinstance(cov, pd.DataFrame):
        return cov
    names = res.model.exog_names
    return pd.DataFrame(np.asarray(cov, dtype=float), index=names, columns=names)


def twoway_cluster_se(
    res_ols,
    groups1: Union[pd.Series, np.ndarray],
    groups2: Union[pd.Series, np.ndarray],
) -> Dict[str, pd.Series]:
    g1 = pd.Series(groups1).astype(str).loc[res_ols._index]
    g2 = pd.Series(groups2).astype(str).loc[res_ols._index]
    g12 = g1 + "×" + g2
    r1 = res_ols.get_robustcov_results(cov_type="cluster", groups=g1)
    r2 = res_ols.get_robustcov_results(cov_type="cluster", groups=g2)
    r12 = res_ols.get_robustcov_results(cov_type="cluster", groups=g12)
    names = res_ols.model.exog_names
    k = len(names)

    def fix(A):
        A = np.asarray(A, dtype=float)
        if A.shape != (k, k):
            A = np.resize(A, (k, k))
        return A

    C1 = fix(r1.cov_params())
    C2 = fix(r2.cov_params())
    C12 = fix(r12.cov_params())
    cov_tw = pd.DataFrame(C1 + C2 - C12, index=names, columns=names)
    se = pd.Series(np.sqrt(np.maximum(np.diag(cov_tw), 0.0)), index=names, name="se")
    params = params_as_series(res_ols)
    tstat = params / se.replace(0.0, np.nan)
    pval = pd.Series(2 * (1 - norm.cdf(np.abs(tstat.values))), index=names, name="p")
    return {"se": se, "t": tstat, "p": pval}


def wild_cluster_bootstrap_pvalue(
    X: np.ndarray,
    y: np.ndarray,
    res,
    groups: Union[pd.Series, np.ndarray],
    param_name: str,
    B: int = 399,
    seed: int = 123,
) -> float:
    rng = np.random.default_rng(seed)
    X = np.asarray(X, dtype=float)
    y = np.asarray(y, dtype=float).reshape(-1)
    groups = pd.Series(groups).astype(str).reset_index(drop=True)
    uniq = groups.unique()
    names = list(res.model.exog_names)
    if param_name not in names:
        return np.nan
    idx = names.index(param_name)
    params = np.asarray(res.params, dtype=float).copy()
    bse = np.asarray(res.bse, dtype=float)
    if idx >= len(params) or idx >= len(bse):
        return np.nan
    if not np.isfinite(bse[idx]) or bse[idx] == 0:
        return np.nan
    t_obs = float(params[idx] / bse[idx])
    b0 = params.copy()
    b0[idx] = 0.0
    yhat0 = X @ b0
    resid_restricted = y - yhat0
    count = 0
    for _ in range(B):
        gsign = pd.Series({g: (1.0 if rng.random() < 0.5 else -1.0) for g in uniq})
        mult = groups.map(gsign).to_numpy(dtype=float)
        y_star = yhat0 + resid_restricted * mult
        try:
            res_b = sm.OLS(y_star, X).fit()
            rb = np.asarray(res_b.bse, dtype=float)
            rp = np.asarray(res_b.params, dtype=float)
            if idx < len(rp) and idx < len(rb):
                if rb[idx] != 0 and np.isfinite(rb[idx]):
                    t_b = float(rp[idx] / rb[idx])
                    if np.isfinite(t_b) and abs(t_b) >= abs(t_obs):
                        count += 1
        except Exception:
            continue
    return (count + 1) / (B + 1)


def safe_wild_boot(
    res,
    df_used: pd.DataFrame,
    param_name: str,
    cluster_col: str = "ym",
    B: Optional[int] = None,
    seed: Optional[int] = None,
) -> float:
    try:
        Xmat = pd.DataFrame(res.model.exog, columns=res.model.exog_names).values
        yvec = pd.Series(res.model.endog).values
        groups = df_used.loc[res._index, cluster_col].astype(str)
        return float(
            wild_cluster_bootstrap_pvalue(
                Xmat,
                yvec,
                res,
                groups,
                param_name,
                B=CFG.WILD_BOOT_B if B is None else B,
                seed=CFG.SEED if seed is None else seed,
            )
        )
    except Exception:
        return np.nan


def report_with_twocluster(
    name: str,
    res,
    df_used: pd.DataFrame,
    key_params: Sequence[str],
    cluster1: str = "bank",
    cluster2: str = "ym",
    wild_for: Optional[Union[str, Sequence[str]]] = None,
) -> None:
    print(f"\n[{name}] two-way cluster results")
    tw = twoway_cluster_se(
        res,
        df_used.loc[res._index, cluster1],
        df_used.loc[res._index, cluster2].astype(str),
    )
    for p in key_params:
        t_tw = float(tw["t"].get(p, np.nan))
        p_tw = float(tw["p"].get(p, np.nan))
        print(f"{p:<25} t_tw={t_tw:8.3f} p_tw={p_tw:0.3f}")
    if wild_for is None:
        wild_list = []
    elif isinstance(wild_for, str):
        wild_list = [wild_for]
    else:
        wild_list = list(wild_for)
    for wf in wild_list:
        p_boot = safe_wild_boot(res, df_used, wf, cluster_col=cluster2)
        print(f"[wild-cluster by {cluster2}] {wf}: p_boot={p_boot:0.3f}")


def make_reg_table(res, var_order: Sequence[str], spec_name: str) -> pd.DataFrame:
    ci = res.conf_int()
    rows = []
    for v in var_order:
        rows.append(
            {
                "spec": spec_name,
                "variable": v,
                "coef": float(params_as_series(res).get(v, np.nan)),
                "se_hac": float(bse_as_series(res).get(v, np.nan)),
                "t_hac": float(pd.Series(res.tvalues, index=res.model.exog_names).get(v, np.nan)),
                "p_hac": float(pvalues_as_series(res).get(v, np.nan)),
                "ci_low": float(ci.loc[v, 0]) if v in ci.index else np.nan,
                "ci_high": float(ci.loc[v, 1]) if v in ci.index else np.nan,
            }
        )
    return pd.DataFrame(rows)


def make_inference_table(
    res,
    df_used: pd.DataFrame,
    spec_name: str,
    params: Sequence[str],
    boot_params: Optional[Sequence[str]] = None,
) -> pd.DataFrame:
    tw = twoway_cluster_se(
        res,
        df_used.loc[res._index, "bank"],
        df_used.loc[res._index, "ym"].astype(str),
    )
    rows = []
    for p in params:
        rows.append(
            {
                "spec": spec_name,
                "param": p,
                "coef": float(params_as_series(res).get(p, np.nan)),
                "se_hac": float(bse_as_series(res).get(p, np.nan)),
                "p_hac": float(pvalues_as_series(res).get(p, np.nan)),
                "se_tw": float(tw["se"].get(p, np.nan)),
                "p_tw": float(tw["p"].get(p, np.nan)),
                "p_boot": float(safe_wild_boot(res, df_used, p)) if (boot_params and p in boot_params) else np.nan,
            }
        )
    return pd.DataFrame(rows)


def combined_effect_table(
    res,
    df_used: pd.DataFrame,
    base_var: str,
    interaction_var: str,
    spec_name: str,
) -> pd.DataFrame:
    b = params_as_series(res)
    cov = _cov_as_df(res)
    if base_var not in cov.index or interaction_var not in cov.columns:
        return pd.DataFrame()
    base = float(b.get(base_var, np.nan))
    inter = float(b.get(interaction_var, np.nan))
    v_base = float(cov.loc[base_var, base_var])
    v_inter = float(cov.loc[interaction_var, interaction_var])
    c_bi = float(cov.loc[base_var, interaction_var])

    beta_non = base
    se_non = np.sqrt(max(v_base, 0.0))
    t_non = beta_non / se_non if se_non > 0 else np.nan
    p_non = 2 * (1 - norm.cdf(abs(t_non))) if np.isfinite(t_non) else np.nan

    beta_elb = base + inter
    var_elb = v_base + v_inter + 2 * c_bi
    se_elb = np.sqrt(max(var_elb, 0.0))
    t_elb = beta_elb / se_elb if se_elb > 0 else np.nan
    p_elb = 2 * (1 - norm.cdf(abs(t_elb))) if np.isfinite(t_elb) else np.nan

    beta_diff = inter
    se_diff = np.sqrt(max(v_inter, 0.0))
    t_diff = beta_diff / se_diff if se_diff > 0 else np.nan
    p_diff = 2 * (1 - norm.cdf(abs(t_diff))) if np.isfinite(t_diff) else np.nan

    return pd.DataFrame(
        [
            {
                "spec": spec_name,
                "effect": f"{base_var} in non-ELB",
                "coef": beta_non,
                "se_hac": se_non,
                "p_hac": p_non,
                "p_boot": np.nan,
            },
            {
                "spec": spec_name,
                "effect": f"{base_var} in ELB",
                "coef": beta_elb,
                "se_hac": se_elb,
                "p_hac": p_elb,
                "p_boot": np.nan,
            },
            {
                "spec": spec_name,
                "effect": "ELB minus non-ELB",
                "coef": beta_diff,
                "se_hac": se_diff,
                "p_hac": p_diff,
                "p_boot": safe_wild_boot(res, df_used, interaction_var),
            },
        ]
    )


def prepare_lowcred_regression_panel(
    panel: pd.DataFrame,
    y_col: str,
    reported_start: str,
) -> pd.DataFrame:
    z = panel.copy()
    z = z.loc[z["date"] >= month_end(reported_start)].copy()
    z["p_share_F"] = z["pi_share"]
    z["p_share_F_x_ELB"] = z["p_share_F"] * z["ELB"]
    z["abs_infl_gap"] = z["inflation_gap"].abs()
    z["miss_streak_24m"] = z.groupby("bank")["abs_infl_gap"].transform(
        lambda s: s.rolling(24, min_periods=6).mean()
    )
    z["abs_infl_gap_x_ELB"] = z["abs_infl_gap"] * z["ELB"]
    z["miss_streak_24m_x_ELB"] = z["miss_streak_24m"] * z["ELB"]
    z["abs_deltai_3_lag"] = z["abs_deltai_3m_lag"]
    z["year"] = z["date"].dt.year
    z["month"] = z["date"].dt.month
    z["ym"] = z["date"].dt.to_period("M")
    if y_col not in z.columns:
        raise KeyError(f"{y_col} not found in regression panel.")
    return z


@dataclass
class PanelCompatConfigLegacy:
    START: str
    END: str
    PI_STAR: float = 2.0
    ELB_EPS: float = 0.10
    ELB_THRESHOLDS: Optional[Dict[str, float]] = None

    def __post_init__(self):
        if self.ELB_THRESHOLDS is None:
            self.ELB_THRESHOLDS = {"ECB": 0.10, "BoE": 0.25, "Fed": 0.25}


def to_monthly_panel_legacy_from_raw(
    raw: Dict[Tuple[str, str], pd.DataFrame],
    cfg_panel: PanelCompatConfigLegacy,
    elb_mode: str = "data_driven",
) -> pd.DataFrame:
    master_me = pd.date_range(month_end(cfg_panel.START), month_end(cfg_panel.END), freq="ME")
    rows = []
    for econ in ["EU", "UK", "US"]:
        bank = BANK_CODE_TO_NAME[econ]
        pol = raw[(econ, "POLICY")]
        inf = raw[(econ, "INF")]
        gdp = raw[(econ, "GDP")]
        m_pol = to_monthly_eom(pol.set_index("date")["value"]).reindex(master_me)
        inf_level_eom = to_monthly_eom(inf.set_index("date")["value"], interpolate_quarterly=True).reindex(master_me)
        m_infl_yoy = infer_yoy_from_INF(inf_level_eom)
        m_gdp = to_monthly_eom(gdp.set_index("date")["value"], interpolate_quarterly=True).reindex(master_me)
        tmp = pd.concat([m_pol, m_infl_yoy, m_gdp], axis=1)
        tmp.columns = ["policy_rate", "infl_yoy", "gdp"]
        tmp["bank"] = bank
        tmp["econ"] = econ
        tmp = tmp.reset_index().rename(columns={"index": "date"})
        rows.append(tmp)
    mon = pd.concat(rows, ignore_index=True).sort_values(["bank", "date"]).reset_index(drop=True)
    mon["gdp"] = pd.to_numeric(mon["gdp"], errors="coerce")
    mon["gdp"] = mon.groupby("bank")["gdp"].transform(lambda s: s.ffill().bfill())
    mon["log_gdp"] = np.log(mon["gdp"].clip(lower=1e-8))
    cyc_list = []
    for bank, g in mon.groupby("bank", sort=False):
        gap = hamilton_output_gap(g.set_index("date")["log_gdp"], k=12)
        tmp = gap.rename("output_gap").reset_index()
        tmp["bank"] = bank
        cyc_list.append(tmp)
    cyc = pd.concat(cyc_list, ignore_index=True)
    mon = mon.merge(cyc, on=["date", "bank"], how="left", validate="one_to_one")
    mon["inflation_gap"] = pd.to_numeric(mon["infl_yoy"], errors="coerce") - cfg_panel.PI_STAR
    mon["year"] = mon["date"].dt.year
    mon["month"] = mon["date"].dt.month
    mon["ym"] = mon["date"].dt.to_period("M")
    mon["crisis10"] = mon["year"].between(2010, 2012).astype(int)
    mon["pandem"] = mon["year"].between(2020, 2021).astype(int)
    mon["QE_FG"] = mon.apply(lambda r: qe_flag(r["bank"], r["year"]), axis=1).astype(int)

    def elb_data_driven_series(s: pd.Series, eps: float) -> pd.Series:
        x = pd.to_numeric(s, errors="coerce")
        roll_min = x.rolling(24, min_periods=24).min()
        return (x <= (roll_min + eps)).astype(int)

    if elb_mode == "data_driven":
        mon["ELB"] = mon.groupby("bank", group_keys=False)["policy_rate"].apply(
            lambda s: elb_data_driven_series(s, cfg_panel.ELB_EPS)
        ).astype(int)
    else:
        mon["ELB"] = mon.apply(
            lambda r: int(r["policy_rate"] <= cfg_panel.ELB_THRESHOLDS.get(str(r["bank"]), 0.25)),
            axis=1,
        ).astype(int)
    mon["policy_rate_lead3"] = mon.groupby("bank")["policy_rate"].shift(-3)
    mon["di"] = mon.groupby("bank")["policy_rate"].diff()
    mon["abs_deltai_3m"] = (mon["policy_rate_lead3"] - mon["policy_rate"]).abs()
    mon["abs_deltai_3m_lag"] = mon.groupby("bank")["abs_deltai_3m"].shift(1)
    mon["abs_pi_gap"] = mon["inflation_gap"].abs()
    mon["abs_y_gap"] = mon["output_gap"].abs()
    den = mon["abs_pi_gap"] + mon["abs_y_gap"]
    mon["pi_share"] = np.where(den > 0, mon["abs_pi_gap"] / den, np.nan)
    return mon


def build_shadow_determinants_panel_legacy(
    raw: Dict[Tuple[str, str], pd.DataFrame],
    cred_eg_raw: pd.DataFrame,
    cfg: JournalConfig,
) -> pd.DataFrame:
    cfg_panel = PanelCompatConfigLegacy(
        START=cfg.SHADOW_START,
        END=cfg.SHADOW_END,
        PI_STAR=cfg.PI_STAR,
        ELB_EPS=cfg.ELB_EPS,
        ELB_THRESHOLDS=cfg.ELB_THRESHOLDS,
    )
    mon = to_monthly_panel_legacy_from_raw(raw, cfg_panel=cfg_panel, elb_mode="data_driven").copy()
    cred_long = cred_eg_raw.copy()
    cred_long.index = pd.to_datetime(cred_long.index)
    cred_long.index = cred_long.index.to_period("M").to_timestamp("M")
    cred_long = cred_long.stack().rename("cred_eg_raw").reset_index()
    cred_long.columns = ["date", "econ", "cred_eg_raw"]
    cred_long["bank"] = cred_long["econ"].map(BANK_CODE_TO_NAME)
    z = mon.merge(
        cred_long[["date", "bank", "cred_eg_raw"]],
        on=["date", "bank"],
        how="left",
        validate="one_to_one",
    ).sort_values(["bank", "date"]).copy()
    z["neg_cred_eg_raw"] = -pd.to_numeric(z["cred_eg_raw"], errors="coerce")
    z["p_share_F"] = z["pi_share"]
    z["p_share_F_x_ELB"] = z["p_share_F"] * z["ELB"]
    z["abs_infl_gap"] = z["inflation_gap"].abs()
    z["miss_streak_24m"] = z.groupby("bank")["abs_infl_gap"].transform(
        lambda s: s.rolling(24, min_periods=6).mean()
    )
    z["abs_infl_gap_x_ELB"] = z["abs_infl_gap"] * z["ELB"]
    z["miss_streak_24m_x_ELB"] = z["miss_streak_24m"] * z["ELB"]
    z["abs_deltai_3_lag"] = z["abs_deltai_3m_lag"]
    z["year"] = z["date"].dt.year
    z["month"] = z["date"].dt.month
    z["ym"] = z["date"].dt.to_period("M")
    z = z[z["date"] >= month_end(cfg.SHADOW_START)].copy()
    return z


def run_workhorse_determinants(
    panel: pd.DataFrame,
    yname: str,
    label_prefix: str,
    heading: str,
    cfg: JournalConfig,
) -> Dict[str, pd.DataFrame]:
    hac_lags = max(cfg.HAC_LAGS_BASE, 2)
    coef_tables = []
    inf_tables = []
    combo_tables = []

    print("\n" + "=" * 100)
    print(heading)
    print("=" * 100)
    print(f"LHS: {yname}")
    print("Interpretation: larger LHS = lower credibility.\n")

    print("-" * 78)
    print("Outcome-based p-share: pooled + ELB interaction")
    print("-" * 78)
    reg_Outcome_based_p_share = panel.copy()
    reg_Outcome_based_p_share = winsorize_df(
        reg_Outcome_based_p_share,
        [c for c in [yname, "p_share_F"] if c in reg_Outcome_based_p_share.columns],
        p=0.01,
    )
    reg_Outcome_based_p_share = reg_Outcome_based_p_share.dropna(
        subset=[yname, "p_share_F", "abs_deltai_3_lag", "ELB", "QE_FG", "crisis10", "pandem"]
    ).copy()

    X_Outcome_based_p_share_pool = [
        "p_share_F",
        "ELB",
        "QE_FG",
        "crisis10",
        "pandem",
        "abs_deltai_3_lag",
    ]
    res_Outcome_based_p_share_pool = fe_ols(
        yname,
        X_Outcome_based_p_share_pool,
        reg_Outcome_based_p_share,
        bank_fe=False,
        bank_year_fe=True,
        month_fe=True,
        hac_lags=hac_lags,
    )
    print("\n[Outcome-based p-share: NO interaction]")
    print(res_Outcome_based_p_share_pool.summary().tables[1])
    report_with_twocluster(
        f"{label_prefix} Outcome-based p-share pooled",
        res_Outcome_based_p_share_pool,
        reg_Outcome_based_p_share,
        ["p_share_F", "ELB"],
        cluster1="bank",
        cluster2="ym",
        wild_for="p_share_F",
    )
    coef_tables.append(
        make_reg_table(
            res_Outcome_based_p_share_pool,
            X_Outcome_based_p_share_pool,
            f"{label_prefix}_Outcome_based_p-share_pooled",
        )
    )
    inf_tables.append(
        make_inference_table(
            res_Outcome_based_p_share_pool,
            reg_Outcome_based_p_share,
            f"{label_prefix}_Outcome_based_p-share_pooled",
            ["p_share_F", "ELB"],
            ["p_share_F"],
        )
    )

    X_Outcome_based_p_share_reg = [
        "p_share_F",
        "ELB",
        "p_share_F_x_ELB",
        "QE_FG",
        "crisis10",
        "pandem",
        "abs_deltai_3_lag",
    ]
    res_Outcome_based_p_share_reg = fe_ols(
        yname,
        X_Outcome_based_p_share_reg,
        reg_Outcome_based_p_share,
        bank_fe=False,
        bank_year_fe=True,
        month_fe=True,
        hac_lags=hac_lags,
    )
    print("\n[Outcome-based p-share with ELB interaction]")
    print(res_Outcome_based_p_share_reg.summary().tables[1])
    report_with_twocluster(
        f"{label_prefix} Outcome-based p-share",
        res_Outcome_based_p_share_reg,
        reg_Outcome_based_p_share,
        ["p_share_F", "p_share_F_x_ELB", "ELB"],
        cluster1="bank",
        cluster2="ym",
        wild_for=["p_share_F", "p_share_F_x_ELB"],
    )
    coef_tables.append(
        make_reg_table(
            res_Outcome_based_p_share_reg,
            X_Outcome_based_p_share_reg,
            f"{label_prefix}_Outcome-based_p-share_regime",
        )
    )
    inf_tables.append(
        make_inference_table(
            res_Outcome_based_p_share_reg,
            reg_Outcome_based_p_share,
            f"{label_prefix}_Outcome_based_p-share_regime",
            ["p_share_F", "p_share_F_x_ELB", "ELB"],
            ["p_share_F", "p_share_F_x_ELB"],
        )
    )
    combo_tables.append(
        combined_effect_table(
            res_Outcome_based_p_share_reg,
            reg_Outcome_based_p_share,
            "p_share_F",
            "p_share_F_x_ELB",
            f"{label_prefix}_Outcome-based_p-share_regime",
        )
    )

    print("\n" + "-" * 78)
    print("Inflation performance: pooled + ELB interactions")
    print("-" * 78)
    reg_Inflation_performance = panel.copy()
    reg_Inflation_performance = winsorize_df(
        reg_Inflation_performance,
        [c for c in [yname, "abs_infl_gap", "miss_streak_24m"] if c in reg_Inflation_performance.columns],
        p=0.01,
    )
    reg_Inflation_performance = reg_Inflation_performance.dropna(
        subset=[yname, "abs_infl_gap", "miss_streak_24m", "output_gap", "ELB", "QE_FG", "crisis10", "pandem"]
    ).copy()

    X_Inflation_performance_pool = [
        "abs_infl_gap",
        "miss_streak_24m",
        "output_gap",
        "ELB",
        "QE_FG",
        "crisis10",
        "pandem",
    ]
    res_Inflation_performance_pool = fe_ols(
        yname,
        X_Inflation_performance_pool,
        reg_Inflation_performance,
        bank_fe=False,
        bank_year_fe=True,
        month_fe=True,
        hac_lags=hac_lags,
    )
    print("\n[Inflation performance pooled: levels-only]")
    print(res_Inflation_performance_pool.summary().tables[1])
    report_with_twocluster(
        f"{label_prefix} Inflation performance pooled",
        res_Inflation_performance_pool,
        reg_Inflation_performance,
        ["abs_infl_gap", "miss_streak_24m"],
        cluster1="bank",
        cluster2="ym",
        wild_for=["abs_infl_gap", "miss_streak_24m"],
    )
    coef_tables.append(
        make_reg_table(
            res_Inflation_performance_pool,
            X_Inflation_performance_pool,
            f"{label_prefix}_Inflation_performance_pooled",
        )
    )
    inf_tables.append(
        make_inference_table(
            res_Inflation_performance_pool,
            reg_Inflation_performance,
            f"{label_prefix}_Inflation_performance_pooled",
            ["abs_infl_gap", "miss_streak_24m"],
            ["abs_infl_gap", "miss_streak_24m"],
        )
    )

    X_Inflation_performance_reg = [
        "abs_infl_gap",
        "abs_infl_gap_x_ELB",
        "miss_streak_24m",
        "miss_streak_24m_x_ELB",
        "output_gap",
        "ELB",
        "QE_FG",
        "crisis10",
        "pandem",
    ]
    res_Inflation_performance_reg = fe_ols(
        yname,
        X_Inflation_performance_reg,
        reg_Inflation_performance,
        bank_fe=False,
        bank_year_fe=True,
        month_fe=True,
        hac_lags=hac_lags,
    )
    print("\n[Inflation performance with ELB interactions]")
    print(res_Inflation_performance_reg.summary().tables[1])
    report_with_twocluster(
        f"{label_prefix} Inflation performance",
        res_Inflation_performance_reg,
        reg_Inflation_performance,
        ["abs_infl_gap", "abs_infl_gap_x_ELB", "miss_streak_24m", "miss_streak_24m_x_ELB"],
        cluster1="bank",
        cluster2="ym",
        wild_for=["abs_infl_gap", "abs_infl_gap_x_ELB", "miss_streak_24m", "miss_streak_24m_x_ELB"],
    )
    coef_tables.append(
        make_reg_table(
            res_Inflation_performance_reg,
            X_Inflation_performance_reg,
            f"{label_prefix}_Inflation_performance_regime",
        )
    )
    inf_tables.append(
        make_inference_table(
            res_Inflation_performance_reg,
            reg_Inflation_performance,
            f"{label_prefix}_Inflation_performance_regime",
            ["abs_infl_gap", "abs_infl_gap_x_ELB", "miss_streak_24m", "miss_streak_24m_x_ELB"],
            ["abs_infl_gap", "abs_infl_gap_x_ELB", "miss_streak_24m", "miss_streak_24m_x_ELB"],
        )
    )
    combo_tables.append(
        combined_effect_table(
            res_Inflation_performance_reg,
            reg_Inflation_performance,
            "abs_infl_gap",
            "abs_infl_gap_x_ELB",
            f"{label_prefix}_Inflation_performance_regime",
        )
    )
    combo_tables.append(
        combined_effect_table(
            res_Inflation_performance_reg,
            reg_Inflation_performance,
            "miss_streak_24m",
            "miss_streak_24m_x_ELB",
            f"{label_prefix}_Inflation_performance_regime",
        )
    )

    print("\n" + "=" * 100)
    print(f"SUMMARY — {label_prefix} DETERMINANTS")
    print("=" * 100)
    b0, se0, p0 = safe_get_triplet(res_Outcome_based_p_share_pool, "p_share_F")
    print(f"Outcome-based p-share pooled: β(p_share_F)={b0:.4f} se={se0:.4f} p={p0:.3f}")
    b0, se0, p0 = safe_get_triplet(res_Outcome_based_p_share_reg, "p_share_F")
    b1, se1, p1 = safe_get_triplet(res_Outcome_based_p_share_reg, "p_share_F_x_ELB")
    print(
        f"Outcome-based p-share: β(p_share_F | ELB=0)={b0:.4f} se={se0:.4f} p={p0:.3f} "
        f"| Δβ at ELB={b1:.4f} se={se1:.4f} p={p1:.3f}"
    )
    b0, se0, p0 = safe_get_triplet(res_Inflation_performance_pool, "abs_infl_gap")
    b1, se1, p1 = safe_get_triplet(res_Inflation_performance_pool, "miss_streak_24m")
    print(
        f"Inflation performance pooled: β(|π-π*|)={b0:.4f} se={se0:.4f} p={p0:.3f} "
        f"| β(streak)={b1:.4f} se={se1:.4f} p={p1:.3f}"
    )
    b0, se0, p0 = safe_get_triplet(res_Inflation_performance_reg, "abs_infl_gap")
    b1, se1, p1 = safe_get_triplet(res_Inflation_performance_reg, "abs_infl_gap_x_ELB")
    print(
        f"Inflation performance: β(|π-π*| | ELB=0)={b0:.4f} se={se0:.4f} p={p0:.3f} "
        f"| Δβ at ELB={b1:.4f} se={se1:.4f} p={p1:.3f}"
    )
    b0, se0, p0 = safe_get_triplet(res_Inflation_performance_reg, "miss_streak_24m")
    b1, se1, p1 = safe_get_triplet(res_Inflation_performance_reg, "miss_streak_24m_x_ELB")
    print(
        f"Inflation performance: β(streak | ELB=0)={b0:.4f} se={se0:.4f} p={p0:.3f} "
        f"| Δβ at ELB={b1:.4f} se={se1:.4f} p={p1:.3f}"
    )

    return {
        "coefficients": pd.concat(coef_tables, ignore_index=True) if coef_tables else pd.DataFrame(),
        "inference": pd.concat(inf_tables, ignore_index=True) if inf_tables else pd.DataFrame(),
        "combined_effects": (
            pd.concat([x for x in combo_tables if x is not None and not x.empty], ignore_index=True)
            if any((x is not None and not x.empty) for x in combo_tables)
            else pd.DataFrame()
        ),
        "regression_samples": {
            "Outcome_based_p_share": reg_Outcome_based_p_share,
            "Inflation_performance": reg_Inflation_performance,
        },
        "models": {
            "Outcome_based_p_share_pool": res_Outcome_based_p_share_pool,
            "Outcome_based_p_share_regime": res_Outcome_based_p_share_reg,
            "Inflation_performance_pool": res_Inflation_performance_pool,
            "Inflation_performance_regime": res_Inflation_performance_reg,
        },
    }


def t5y5y_monthly_eom(t5y5y_daily: pd.Series) -> pd.Series:
    s = pd.Series(t5y5y_daily).copy()
    if not isinstance(s.index, pd.DatetimeIndex):
        s.index = pd.to_datetime(s.index, errors="coerce")
    s = pd.to_numeric(s, errors="coerce")
    s = s.dropna().sort_index()
    return s.resample("ME").last().rename("t5y5y")


def run_us_t5y5y_validation(
    panel: pd.DataFrame,
    yvar: str,
    label: str,
    t5y5y_daily: pd.Series,
    cfg: JournalConfig,
    val_start: str = "2012-01-31",
    val_end: Optional[str] = None,
) -> Dict[str, pd.DataFrame]:
    if val_end is None:
        val_end = cfg.END
    t5y5y = t5y5y_monthly_eom(t5y5y_daily)
    fed = panel[panel["bank"] == "Fed"].copy()
    fed = fed.dropna(subset=["date", yvar]).sort_values("date")
    fed = fed.merge(t5y5y.reset_index(), on="date", how="left", validate="one_to_one")
    fed = fed[
        (fed["date"] >= month_end(val_start))
        & (fed["date"] <= month_end(val_end))
    ].copy()
    fed["abs_anch_dev"] = (fed["t5y5y"] - cfg.PI_STAR).abs()
    fed["downside_anch_gap"] = (cfg.PI_STAR - fed["t5y5y"]).clip(lower=0)
    joint = fed.dropna(
        subset=[yvar, "t5y5y", "abs_anch_dev", "downside_anch_gap"]
    ).copy()
    joint = joint.sort_values("date").reset_index(drop=True)

    correlations = pd.DataFrame(
        [
            {
                "label": label,
                "sample_start": joint["date"].min(),
                "sample_end": joint["date"].max(),
                "N": int(len(joint)),
                "corr_y_t5y5y_pearson": float(joint[yvar].corr(joint["t5y5y"], method="pearson")),
                "corr_y_t5y5y_spearman": float(joint[yvar].corr(joint["t5y5y"], method="spearman")),
                "corr_y_absdev_pearson": float(joint[yvar].corr(joint["abs_anch_dev"], method="pearson")),
                "corr_y_absdev_spearman": float(joint[yvar].corr(joint["abs_anch_dev"], method="spearman")),
                "corr_y_downside_gap_pearson": float(joint[yvar].corr(joint["downside_anch_gap"], method="pearson")),
                "corr_y_downside_gap_spearman": float(joint[yvar].corr(joint["downside_anch_gap"], method="spearman")),
            }
        ]
    )

    reg = joint.copy()
    reg["year"] = reg["date"].dt.year
    reg["month"] = reg["date"].dt.month

    def run_downside_reg(spec_name: str, year_fe: bool, month_fe: bool) -> Dict[str, object]:
        X = pd.DataFrame(
            {
                "const": 1.0,
                "downside_anch_gap": reg["downside_anch_gap"].astype(float),
            },
            index=reg.index,
        )
        if year_fe:
            X = pd.concat(
                [X, pd.get_dummies(reg["year"], prefix="y", drop_first=True).astype(float)],
                axis=1,
            )
        if month_fe:
            X = pd.concat(
                [X, pd.get_dummies(reg["month"], prefix="m", drop_first=True).astype(float)],
                axis=1,
            )
        y = reg[yvar].astype(float)
        res = sm.OLS(y, X).fit(
            cov_type="HAC",
            cov_kwds={"maxlags": cfg.HAC_LAGS_BASE},
        )
        ci = res.conf_int()
        return {
            "spec": spec_name,
            "variable": "downside_anch_gap",
            "coef": float(res.params.get("downside_anch_gap", np.nan)),
            "se_hac": float(res.bse.get("downside_anch_gap", np.nan)),
            "t_hac": float(res.tvalues.get("downside_anch_gap", np.nan)),
            "p_hac": float(res.pvalues.get("downside_anch_gap", np.nan)),
            "ci_low": float(ci.loc["downside_anch_gap", 0]) if "downside_anch_gap" in ci.index else np.nan,
            "ci_high": float(ci.loc["downside_anch_gap", 1]) if "downside_anch_gap" in ci.index else np.nan,
            "N": int(res.nobs),
            "sample_start": reg["date"].min(),
            "sample_end": reg["date"].max(),
            "year_FE": year_fe,
            "month_FE": month_fe,
        }

    regression = pd.DataFrame(
        [
            run_downside_reg("no_FE", year_fe=False, month_fe=False),
            run_downside_reg("month_FE", year_fe=False, month_fe=True),
        ]
    )

    baseline_row = regression.loc[regression["spec"] == "month_FE"].iloc[0]
    summary = pd.DataFrame(
        [
            {
                "label": label,
                "validation_object": "downside_anch_gap = max(2 - 5y5y, 0)",
                "baseline_spec": "month_FE",
                "higher_is_worse": True,
                "beta_downside_anch_gap": float(baseline_row["coef"]),
                "se_hac": float(baseline_row["se_hac"]),
                "p_hac": float(baseline_row["p_hac"]),
                "n_joint": int(baseline_row["N"]),
                "sample_start": baseline_row["sample_start"],
                "sample_end": baseline_row["sample_end"],
            }
        ]
    )

    return {
        "correlations": correlations,
        "regression": regression,
        "summary": summary,
        "validation_data": joint,
    }



def build_symmetric_validation_panel(
    ap3_panel_reported: pd.DataFrame,
    shadow_bundle: Dict[str, object],
) -> pd.DataFrame:
    ap3 = ap3_panel_reported[["date", "bank", "AP3_raw_lowcred"]].copy()
    ap3["date"] = pd.to_datetime(ap3["date"]).dt.to_period("M").dt.to_timestamp("M")
    ap3["econ"] = ap3["bank"].map(BANK_NAME_TO_CODE)
    ap3 = ap3.rename(columns={"AP3_raw_lowcred": "AP3_LOW"})

    def _shadow_to_long(raw_df: pd.DataFrame, stem: str) -> pd.DataFrame:
        z = raw_df.copy()
        z.index = pd.to_datetime(z.index).to_period("M").to_timestamp("M")
        long = z.stack().rename(f"{stem}_HIGH").reset_index()
        long.columns = ["date", "econ", f"{stem}_HIGH"]
        long["bank"] = long["econ"].map(BANK_CODE_TO_NAME)
        long[f"{stem}_LOW"] = -pd.to_numeric(long[f"{stem}_HIGH"], errors="coerce")
        return long[["date", "econ", "bank", f"{stem}_HIGH", f"{stem}_LOW"]]

    eg = _shadow_to_long(shadow_bundle["cred_eg_raw"], "SHADOW_EG")
    gjr = _shadow_to_long(shadow_bundle["cred_gjr_raw"], "SHADOW_GJR")

    out = ap3.merge(
        eg,
        on=["date", "econ", "bank"],
        how="outer",
        validate="one_to_one",
    )
    out = out.merge(
        gjr,
        on=["date", "econ", "bank"],
        how="outer",
        validate="one_to_one",
    )
    return out.sort_values(["bank", "date"]).reset_index(drop=True)


def run_us_t5y5y_validation_symmetric(
    proxy_panel: pd.DataFrame,
    t5y5y_daily: pd.Series,
    cfg: JournalConfig,
    val_start: str = "2012-01-31",
    val_end: Optional[str] = None,
    proxy_cols: Sequence[str] = ("AP3_LOW", "SHADOW_EG_LOW"),
) -> Dict[str, pd.DataFrame]:
    if val_end is None:
        val_end = cfg.END

    t5 = t5y5y_monthly_eom(t5y5y_daily).reset_index()
    t5["date"] = pd.to_datetime(t5["date"]).dt.to_period("M").dt.to_timestamp("M")

    fed = proxy_panel.loc[proxy_panel["bank"] == "Fed"].copy()
    fed["date"] = pd.to_datetime(fed["date"]).dt.to_period("M").dt.to_timestamp("M")
    df = fed.merge(t5, on="date", how="left", validate="one_to_one")
    df = df[
        (df["date"] >= month_end(val_start))
        & (df["date"] <= month_end(val_end))
    ].copy()
    df["abs_anch_dev"] = (df["t5y5y"] - cfg.PI_STAR).abs()
    df["downside_anch_gap"] = (cfg.PI_STAR - df["t5y5y"]).clip(lower=0)
    df["year"] = df["date"].dt.year
    df["month"] = df["date"].dt.month

    corr_rows = []
    reg_rows = []

    def _run_one_reg(data: pd.DataFrame, proxy: str, spec_name: str, year_fe: bool, month_fe: bool) -> Dict[str, object]:
        X = pd.DataFrame(
            {
                "const": 1.0,
                "downside_anch_gap": data["downside_anch_gap"].astype(float),
            },
            index=data.index,
        )
        if year_fe:
            X = pd.concat(
                [X, pd.get_dummies(data["year"], prefix="y", drop_first=True).astype(float)],
                axis=1,
            )
        if month_fe:
            X = pd.concat(
                [X, pd.get_dummies(data["month"], prefix="m", drop_first=True).astype(float)],
                axis=1,
            )
        y = data[proxy].astype(float)
        res = sm.OLS(y, X).fit(
            cov_type="HAC",
            cov_kwds={"maxlags": cfg.HAC_LAGS_BASE},
        )
        ci = res.conf_int()
        return {
            "proxy": proxy,
            "spec": spec_name,
            "variable": "downside_anch_gap",
            "coef": float(res.params.get("downside_anch_gap", np.nan)),
            "se_hac": float(res.bse.get("downside_anch_gap", np.nan)),
            "t_hac": float(res.tvalues.get("downside_anch_gap", np.nan)),
            "p_hac": float(res.pvalues.get("downside_anch_gap", np.nan)),
            "ci_low": float(ci.loc["downside_anch_gap", 0]) if "downside_anch_gap" in ci.index else np.nan,
            "ci_high": float(ci.loc["downside_anch_gap", 1]) if "downside_anch_gap" in ci.index else np.nan,
            "N": int(res.nobs),
            "sample_start": data["date"].min(),
            "sample_end": data["date"].max(),
            "year_FE": bool(year_fe),
            "month_FE": bool(month_fe),
            "higher_proxy_is_lower_credibility": True,
        }

    for proxy in proxy_cols:
        if proxy not in df.columns:
            continue
        joint = df.dropna(
            subset=[proxy, "t5y5y", "abs_anch_dev", "downside_anch_gap"]
        ).sort_values("date").copy()
        if joint.empty:
            continue

        corr_rows.append(
            {
                "proxy": proxy,
                "sample_start": joint["date"].min(),
                "sample_end": joint["date"].max(),
                "N": int(len(joint)),
                "corr_proxy_t5y5y_pearson": float(joint[proxy].corr(joint["t5y5y"], method="pearson")),
                "corr_proxy_t5y5y_spearman": float(joint[proxy].corr(joint["t5y5y"], method="spearman")),
                "corr_proxy_absdev_pearson": float(joint[proxy].corr(joint["abs_anch_dev"], method="pearson")),
                "corr_proxy_absdev_spearman": float(joint[proxy].corr(joint["abs_anch_dev"], method="spearman")),
                "corr_proxy_downside_gap_pearson": float(joint[proxy].corr(joint["downside_anch_gap"], method="pearson")),
                "corr_proxy_downside_gap_spearman": float(joint[proxy].corr(joint["downside_anch_gap"], method="spearman")),
                "higher_proxy_is_lower_credibility": True,
            }
        )

        reg_rows.extend(
            [
                _run_one_reg(joint, proxy, "no_FE", year_fe=False, month_fe=False),
                _run_one_reg(joint, proxy, "month_FE", year_fe=False, month_fe=True),
            ]
        )

    regression = pd.DataFrame(reg_rows)
    if not regression.empty:
        summary = regression.loc[regression["spec"] == "month_FE"].copy()
        summary = summary.rename(
            columns={
                "coef": "beta_downside_anch_gap",
                "N": "n_joint",
            }
        )
        summary["validation_object"] = "downside_anch_gap = max(2 - 5y5y, 0)"
        summary["baseline_spec"] = "month_FE"
    else:
        summary = pd.DataFrame()

    return {
        "validation_data": df,
        "correlations": pd.DataFrame(corr_rows),
        "regression": regression,
        "summary": summary,
    }


def build_sacrifice_ratio_validation_base(
    shadow_wx_panel: pd.DataFrame,
    cfg: JournalConfig,
) -> pd.DataFrame:
    sr_rows = []
    for e in ["US", "EU", "UK"]:
        sub = shadow_wx_panel.xs(e, level="econ").sort_index()
        gap = hamilton_gap_from_gdp_legacy(sub["GDP"], k=cfg.HAMILTON_K)
        pi = sub["PI"]
        srH = sacrifice_ratio_legacy(
            pi,
            gap,
            H=cfg.SR_H,
            min_abs_dpi=cfg.MIN_ABS_DPI,
            winsor_pcts=cfg.WINSOR_PCTS,
        )
        sr_rows.append(pd.DataFrame({"date": srH.index, "econ": e, "SR": srH.values}))

    sr_df = pd.concat(sr_rows, axis=0)
    if cfg.LEGACY_SR_LAG_AFTER_DROP:
        sr_df = sr_df.dropna(subset=["SR"])

    sr_df["date"] = pd.to_datetime(sr_df["date"]).dt.to_period("M").dt.to_timestamp(how="start")
    sr_df["bank"] = sr_df["econ"].map(BANK_CODE_TO_NAME)
    sr_df = sr_df.sort_values(["econ", "date"]).reset_index(drop=True)
    return sr_df


def _run_sr_reg_hac_general(
    sr_panel_long: pd.DataFrame,
    proxy_col: str,
    H: int,
    L: int,
    analysis_start: Optional[str] = None,
):
    df = sr_panel_long.copy().sort_values(["econ", "date"])
    lag_name = f"{proxy_col}_lag{L}"
    df[lag_name] = df.groupby("econ")[proxy_col].shift(L)
    if analysis_start is not None:
        df = df[df["date"] >= pd.Timestamp(analysis_start)].copy()

    use = df[["SR", lag_name, "econ"]].dropna().copy()
    D = pd.get_dummies(use["econ"], prefix="ent", drop_first=True, dtype=int)
    X = pd.concat([use[[lag_name]], D], axis=1)
    X = sm.add_constant(X, has_constant="add")
    y = use["SR"]
    res_hac = sm.OLS(y, X).fit(cov_type="HAC", cov_kwds={"maxlags": H})
    return res_hac, int(res_hac.nobs), use, lag_name


def run_sacrifice_ratio_validation_symmetric(
    shadow_wx_panel: pd.DataFrame,
    proxy_panel: pd.DataFrame,
    cfg: JournalConfig,
    proxy_cols: Sequence[str] = ("AP3_LOW", "SHADOW_EG_LOW"),
) -> Dict[str, object]:
    sr_df = build_sacrifice_ratio_validation_base(shadow_wx_panel, cfg)

    px = proxy_panel.copy()
    px["date"] = pd.to_datetime(px["date"]).dt.to_period("M").dt.to_timestamp(how="start")
    keep = ["date", "econ", "bank"] + [c for c in proxy_cols if c in px.columns]
    px = px[keep].drop_duplicates(subset=["date", "econ", "bank"])

    sr_df = sr_df.merge(
        px,
        on=["date", "econ", "bank"],
        how="left",
        validate="one_to_one",
    ).sort_values(["econ", "date"]).reset_index(drop=True)

    H = cfg.SR_H
    L = cfg.SR_LAG
    rows = []
    models = {}
    samples = {}

    for proxy in proxy_cols:
        if proxy not in sr_df.columns:
            continue
        try:
            res, N, use, lag_name = _run_sr_reg_hac_general(
                sr_df,
                proxy_col=proxy,
                H=H,
                L=L,
                analysis_start=cfg.SHADOW_START,
            )
            b, se, p = coef_line(res, lag_name)
            ci = res.conf_int()
            rows.append(
                {
                    "proxy": proxy,
                    "model": f"SR_H{H} ~ {proxy}(t-{L}) + country dummies",
                    "lag": int(L),
                    "coef": b,
                    "se_hac": se,
                    "p_hac": p,
                    "ci_low": float(ci.loc[lag_name, 0]) if lag_name in ci.index else np.nan,
                    "ci_high": float(ci.loc[lag_name, 1]) if lag_name in ci.index else np.nan,
                    "N": int(N),
                    "HAC_lags": int(H),
                    "higher_proxy_is_lower_credibility": True,
                    "legacy_sr_lag_after_drop": bool(cfg.LEGACY_SR_LAG_AFTER_DROP),
                }
            )
            models[proxy] = res
            samples[proxy] = use
        except Exception as e:
            rows.append(
                {
                    "proxy": proxy,
                    "model": f"SR_H{H} ~ {proxy}(t-{L}) + country dummies",
                    "lag": int(L),
                    "coef": np.nan,
                    "se_hac": np.nan,
                    "p_hac": np.nan,
                    "ci_low": np.nan,
                    "ci_high": np.nan,
                    "N": 0,
                    "HAC_lags": int(H),
                    "higher_proxy_is_lower_credibility": True,
                    "legacy_sr_lag_after_drop": bool(cfg.LEGACY_SR_LAG_AFTER_DROP),
                    "error": str(e),
                }
            )

    return {
        "sr_df": sr_df,
        "summary": pd.DataFrame(rows),
        "models": models,
        "samples": samples,
    }


def run_proxy_noninterchangeability_symmetric(
    proxy_panel: pd.DataFrame,
    proxy_cols: Sequence[str] = ("AP3_LOW", "SHADOW_EG_LOW"),
) -> pd.DataFrame:
    rows = []
    available = [c for c in proxy_cols if c in proxy_panel.columns]
    pairs = []
    for i, a in enumerate(available):
        for b in available[i + 1:]:
            pairs.append((a, b))

    for bank, g in proxy_panel.groupby("bank"):
        for a, b in pairs:
            use = g[[a, b]].dropna()
            rows.append(
                {
                    "bank": bank,
                    "proxy_a": a,
                    "proxy_b": b,
                    "N": int(len(use)),
                    "pearson": float(use[a].corr(use[b], method="pearson")) if len(use) > 5 else np.nan,
                    "spearman": float(use[a].corr(use[b], method="spearman")) if len(use) > 5 else np.nan,
                }
            )

    for a, b in pairs:
        use = proxy_panel[[a, b]].dropna()
        rows.append(
            {
                "bank": "pooled",
                "proxy_a": a,
                "proxy_b": b,
                "N": int(len(use)),
                "pearson": float(use[a].corr(use[b], method="pearson")) if len(use) > 5 else np.nan,
                "spearman": float(use[a].corr(use[b], method="spearman")) if len(use) > 5 else np.nan,
            }
        )

    return pd.DataFrame(rows)


def maybe_write_results(results: Dict[str, object], cfg: JournalConfig) -> None:
    if not cfg.WRITE_RESULTS_TO_FILES:
        return
    outdir = Path(cfg.RESULTS_DIR)
    outdir.mkdir(parents=True, exist_ok=True)
    tables = {
        "workbook_audit": results.get("workbook_audit"),
        "macro_backbone_audit": results.get("macro_backbone_audit"),
        "ap3_rf_window_cv": results["ap3"].get("rf_window_cv"),
        "us_5y5y_correlations": results["us_val_ap3"].get("correlations"),
        "us_5y5y_regression": results["us_val_ap3"].get("regression"),
        "us_5y5y_summary": results["us_val_ap3"].get("summary"),
        "ap3_det_coefficients": results["det_ap3"].get("coefficients"),
        "ap3_det_inference": results["det_ap3"].get("inference"),
        "ap3_det_combined_effects": results["det_ap3"].get("combined_effects"),
        "shadow_det_coefficients": results["det_shadow"].get("coefficients"),
        "shadow_det_inference": results["det_shadow"].get("inference"),
        "shadow_det_combined_effects": results["det_shadow"].get("combined_effects"),
        "egarch_meta": results["shadow"].get("eg_meta"),
        "gjr_meta": results["shadow"].get("gjr_meta"),
        "gjr_corr_by_econ": results["shadow"].get("gjr_corr_by_econ"),
        "validation_proxy_panel": results.get("validation_proxy_panel"),
        "us_5y5y_symmetric_correlations": results["us_val_symmetric"].get("correlations"),
        "us_5y5y_symmetric_regression": results["us_val_symmetric"].get("regression"),
        "us_5y5y_symmetric_summary": results["us_val_symmetric"].get("summary"),
        "sacrifice_ratio_symmetric_summary": results["sr_val_symmetric"].get("summary"),
        "proxy_noninterchangeability": results.get("proxy_noninterchangeability"),
    }
    for name, df in tables.items():
        if isinstance(df, pd.DataFrame):
            df.to_csv(outdir / f"{name}.csv", index=False)


def replication_assertions(results: Dict[str, object], cfg: JournalConfig) -> None:
    ap3_panel = results["ap3"]["panel_reported"]
    shadow = results["shadow"]
    assert ap3_panel["AP3_raw_lowcred"].notna().sum() > 0, "AP3_raw_lowcred has no observations."
    assert shadow["cred_eg_raw"].notna().sum().sum() > 0, "Shadow raw credibility has no observations."
    assert pd.DatetimeIndex(shadow["cred_eg_raw"].index).is_month_start.all(), "Shadow credibility is not month-start."
    assert pd.DatetimeIndex(results["shadow_wide"].index).is_month_start.all(), "Shadow-rate panel is not month-start."
    if cfg.LEGACY_SHADOW_FORCE_AR0:
        ar_lags = shadow["eg_meta"]["ar_lags"].astype(int)
        assert ar_lags.eq(0).all(), "Legacy forced AR(0) failed."
    assert results["us_val_ap3"]["correlations"]["N"].iloc[0] > 0, "US 5y5y validation has no joint observations."
    if "us_val_symmetric" in results:
        assert not results["us_val_symmetric"]["regression"].empty, "Symmetric 5y5y validation has no regressions."
    if "sr_val_symmetric" in results:
        assert not results["sr_val_symmetric"]["summary"].empty, "Symmetric sacrifice-ratio validation has no regressions."


def run_all(cfg: JournalConfig = CFG) -> Dict[str, object]:
    global CFG
    CFG = cfg
    set_seeds(cfg.SEED)
    data_path = resolve_input_file(cfg.DATA_FILE, cfg.ROOT_LOCAL)

    show_heading("Single-workbook data audit")
    audit = freq_audit(data_path)
    print(audit.to_string(index=False))

    raw, t5y5y_daily = load_unified_raw(data_path)

    common_panel = build_common_macro_panel(raw, cfg=cfg, elb_mode="data_driven")
    show_table("AP3 macro backbone audit", macro_backbone_audit(common_panel))

    ap3_bundle = build_ap3_measure(common_panel, cfg)
    ap3_panel_for_det = prepare_lowcred_regression_panel(
        ap3_bundle["panel_reported"],
        y_col="AP3_raw_lowcred",
        reported_start=cfg.REPORTED_START,
    )
    det_ap3 = run_workhorse_determinants(
        ap3_panel_for_det,
        yname="AP3_raw_lowcred",
        label_prefix="AP3_RAW",
        heading="AP(3) DETERMINANTS — RAW LOW-CREDIBILITY",
        cfg=cfg,
    )

    us_val_ap3 = run_us_t5y5y_validation(
        ap3_bundle["panel_reported"].rename(columns={"AP3_raw_lowcred": "AP3_RAW"}),
        yvar="AP3_RAW",
        label="AP(3) raw low-credibility",
        t5y5y_daily=t5y5y_daily,
        cfg=cfg,
    )

    shadow_wide = build_shadow_rate_wide(raw, cfg=cfg)
    shadow_bundle = run_shadow_rate_measure(shadow_wide, cfg)

    shadow_wx_panel = prep_shadow_panel_legacy(raw, cfg)
    sr_validation = run_sacrifice_ratio_validation_legacy(
        shadow_wx_panel=shadow_wx_panel,
        cred_eg_raw=shadow_bundle["cred_eg_raw"],
        cfg=cfg,
    )

    validation_proxy_panel = build_symmetric_validation_panel(
        ap3_panel_reported=ap3_bundle["panel_reported"],
        shadow_bundle=shadow_bundle,
    )
    us_val_symmetric = run_us_t5y5y_validation_symmetric(
        proxy_panel=validation_proxy_panel,
        t5y5y_daily=t5y5y_daily,
        cfg=cfg,
    )
    sr_val_symmetric = run_sacrifice_ratio_validation_symmetric(
        shadow_wx_panel=shadow_wx_panel,
        proxy_panel=validation_proxy_panel,
        cfg=cfg,
    )
    proxy_noninterchangeability = run_proxy_noninterchangeability_symmetric(
        validation_proxy_panel
    )

    shadow_panel_for_det = build_shadow_determinants_panel_legacy(
        raw=raw,
        cred_eg_raw=shadow_bundle["cred_eg_raw"],
        cfg=cfg,
    )
    det_shadow = run_workhorse_determinants(
        shadow_panel_for_det,
        yname="neg_cred_eg_raw",
        label_prefix="SHADOW_RAW",
        heading="SHADOW-RATE CREDIBILITY DETERMINANTS — NEGATIVE RAW EGARCH CREDIBILITY",
        cfg=cfg,
    )

    results = {
        "data_path": data_path,
        "workbook_audit": audit,
        "raw": raw,
        "t5y5y_daily": t5y5y_daily,
        "common_panel": common_panel,
        "macro_backbone_audit": macro_backbone_audit(common_panel),
        "shadow_wide": shadow_wide,
        "shadow_wx_panel": shadow_wx_panel,
        "ap3": ap3_bundle,
        "det_ap3": det_ap3,
        "us_val_ap3": us_val_ap3,
        "shadow": shadow_bundle,
        "sacrifice_ratio": sr_validation,
        "validation_proxy_panel": validation_proxy_panel,
        "us_val_symmetric": us_val_symmetric,
        "sr_val_symmetric": sr_val_symmetric,
        "proxy_noninterchangeability": proxy_noninterchangeability,
        "shadow_det_panel": shadow_panel_for_det,
        "det_shadow": det_shadow,
    }

    replication_assertions(results, cfg)
    maybe_write_results(results, cfg)
    return results


def present_results(results: Dict[str, object]) -> None:
    print(f"\nData workbook: {results['data_path']}")
    print(f"Best AP(3) rolling RF window: {results['ap3']['best_rf_window']}")
    show_table("AP(3) rolling-window CV", results["ap3"]["rf_window_cv"])
    show_table("AP(3) scale stats", pd.DataFrame([results["ap3"]["scale_stats"]]))
    show_table("US 5y5y validation: correlations", results["us_val_ap3"]["correlations"])
    show_table("US 5y5y validation: downside-gap regressions", results["us_val_ap3"]["regression"])
    show_table("US 5y5y validation: baseline summary", results["us_val_ap3"]["summary"])
    if "us_val_symmetric" in results:
        show_table(
            "SYMMETRIC validation: US 5y5y correlations",
            results["us_val_symmetric"]["correlations"],
        )
        show_table(
            "SYMMETRIC validation: US 5y5y downside-gap regressions",
            results["us_val_symmetric"]["regression"],
        )
        show_table(
            "SYMMETRIC validation: US 5y5y baseline summaries",
            results["us_val_symmetric"]["summary"],
        )
    show_table("Shadow EGARCH meta", results["shadow"]["eg_meta"])
    show_table("Shadow anchor stats", pd.DataFrame([results["shadow"]["anchor_stats"]]))
    show_table("Shadow EGARCH vs GJR robustness", results["shadow"]["gjr_corr_by_econ"])
    sr_res = results["sacrifice_ratio"]["res_hac_eg"]
    xname = f"cred_lag{CFG.SR_LAG}"
    b, se, p = coef_line(sr_res, xname)
    show_table(
        "Sacrifice-ratio validation summary",
        pd.DataFrame(
            [
                {
                    "model": f"SR_H{CFG.SR_H} ~ {xname} + country dummies",
                    "coef": b,
                    "se_hac": se,
                    "p_hac": p,
                    "N": int(sr_res.nobs),
                    "HAC_lags": CFG.SR_H,
                }
            ]
        ),
    )
    if "sr_val_symmetric" in results:
        show_table(
            "SYMMETRIC validation: sacrifice-ratio regressions",
            results["sr_val_symmetric"]["summary"],
        )
    if "proxy_noninterchangeability" in results:
        show_table(
            "SYMMETRIC validation: proxy non-interchangeability",
            results["proxy_noninterchangeability"],
        )
    show_table("AP3 determinant coefficients", results["det_ap3"]["coefficients"])
    show_table("Shadow determinant coefficients", results["det_shadow"]["coefficients"])


if __name__ == "__main__":
    results = run_all(CFG)
    present_results(results)
```
