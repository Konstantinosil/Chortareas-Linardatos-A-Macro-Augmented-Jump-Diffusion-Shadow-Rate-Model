```python
import sys, subprocess, importlib.util
_needed = []
for _pkg, _pip_name in [("filterpy", "filterpy==1.4.5"), ("openpyxl", "openpyxl"), ("tqdm", "tqdm")]:
    if importlib.util.find_spec(_pkg) is None:
        _needed.append(_pip_name)
if _needed:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", *_needed])


import numpy as np, pandas as pd, matplotlib.pyplot as plt
from pathlib import Path; from scipy.optimize import minimize
from filterpy.kalman import KalmanFilter
from sklearn.decomposition import PCA; from sklearn.preprocessing import StandardScaler
from statsmodels.regression.linear_model import OLS; from statsmodels.tools.tools import add_constant
from statsmodels.tsa.stattools import adfuller
from statsmodels.stats.diagnostic import acorr_ljungbox, het_arch
from statsmodels.tsa.ar_model import AutoReg
from IPython.display import display

SEED = 42
rng  = np.random.default_rng(SEED);  np.random.seed(SEED)

def find_input_file(filename):
    """
    Locate replication-package input files in a portable way.

    The notebooks can be run from:
    - the same folder as the input files,
    - a local data/raw folder,
    - Google Colab /content,
    - this sandbox /mnt/data.
    """
    filename = str(filename)
    search_dirs = [
        Path.cwd(),
        Path.cwd() / "data",
        Path.cwd() / "data" / "raw",
        Path.cwd().parent,
        Path("/content"),
        Path("/content") / "data",
        Path("/content") / "data" / "raw",
        Path("/mnt/data"),
    ]

    for d in search_dirs:
        p = d / filename
        if p.exists():
            return p

    for root in [Path.cwd(), Path("/content"), Path("/mnt/data")]:
        if root.exists():
            matches = list(root.glob(f"**/{filename}"))
            if matches:
                return matches[0]

    raise FileNotFoundError(
        f"Could not find '{filename}'. Put it in the notebook working directory, "
        "a data/raw folder, or /content in Colab."
    )

OUTPUT_DIR = Path("replication_outputs_us")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

DATA = find_input_file("Dataset United States.xlsx")
xls  = pd.ExcelFile(DATA)
print(f"Using master dataset: {DATA.resolve()}")
print(f"Writing outputs to: {OUTPUT_DIR.resolve()}")

def _prep(sheet, cols):
    df = pd.read_excel(xls, sheet, header=1)
    df.columns = cols
    df["Date"] = pd.to_datetime(df["Date"], dayfirst=True, errors="coerce")
    return df.dropna(subset=["Date"]).set_index("Date")

ser = {
    "SR" : _prep("Shadow Rates US", ["Date","SR US"]),
    "INF": _prep("Inflation Data US", ["Date","INF US"]),
    "UN" : _prep("Unemployment Data US", ["Date","UNEM US"]),
    "CDS": _prep("Sovereign CDS Spreads US", ["Date","CDS US"]),
    "GDP": _prep("Real GDP US", ["Date","GDP US"]),
    "EX" : _prep("Exchange Rates US", ["Date","EX US"]),
    "3M" : _prep("3M",  ["Date","3M US"]),
    "6M" : _prep("6M",  ["Date","6M US"]),
    "1Y" : _prep("1Y",  ["Date","1Y US"]),
    "2Y" : _prep("2Y",  ["Date","2Y US"]),
    "5Y" : _prep("5Y",  ["Date","5Y US"]),
    "7Y" : _prep("7Y",  ["Date","7Y US"]),
    "10Y": _prep("10Y", ["Date","10Y US"]),
    "30Y": _prep("30Y", ["Date","30Y US"]),
}

df = pd.concat(ser.values(), axis=1).interpolate("linear")
macro = ["INF US","UNEM US","EX US","CDS US","GDP US"]
ylds  = ["3M US","6M US","1Y US","2Y US","5Y US","7Y US","10Y US","30Y US"]
df    = df.dropna(subset=["SR US"] + macro + ylds)

y_df, m_df = df[ylds].copy(), df[macro].copy()

var_list = ylds + macro
tbl_station, tbl_ac, tbl_arch = [], [], []

for v in var_list:
    s0 = df[v].dropna()
    adf0 = adfuller(s0, autolag='AIC'); stat0, p0 = adf0[0], adf0[1]

    if p0 < .05:
        tbl_station.append([v, stat0, p0, 'Stat']);  s_final = s0
    else:
        s1 = s0.diff().dropna()
        adf1 = adfuller(s1, autolag='AIC')
        tbl_station += [[f'{v}(orig)',stat0,p0,'NOT stat'],
                        [f'{v}(Δ1)',adf1[0],adf1[1],
                         'Stat' if adf1[1]<.05 else 'NOT']]
        s_final = s1

    lb_p   = acorr_ljungbox(s_final, lags=[10], return_df=True)['lb_pvalue'].iloc[0]
    arch_p = het_arch(s_final, maxlag=10)[1]
    tbl_ac.append([v, lb_p,   'Sig' if lb_p<.05  else 'OK'])
    tbl_arch.append([v, arch_p,'Sig' if arch_p<.05 else 'OK'])

print("\n=== ADF ===");   display(pd.DataFrame(tbl_station, columns=['Var','ADF','p','Res']))
print("\n=== LB  ===");  display(pd.DataFrame(tbl_ac,      columns=['Var','p','AC?']))
print("\n=== ARCH ==="); display(pd.DataFrame(tbl_arch,    columns=['Var','p','ARCH?']))

if y_df.abs().max().max() > 20: y_df /= 100
low, upp = y_df.mean()-3*y_df.std(), y_df.mean()+3*y_df.std()
y_df = y_df.clip(lower=low, upper=upp, axis=1)
m_z  = pd.DataFrame(StandardScaler().fit_transform(m_df),
                    index=m_df.index, columns=m_df.columns)

def _beta(y,z):
    X = add_constant(z); return {c: OLS(y[c],X).fit().params.values for c in y}

def _pca(y): return PCA(1,random_state=SEED).fit_transform(y).ravel()

def _sim(n,r0,z,β,ψ,lb=-2):
    κ,σ,λ,μ,s,θ0,θ1,θ2 = ψ; z0,z1 = 240,1992
    S=np.empty(n); S[0]=r0
    for t in range(1,n):
        θ = θ0 if t<z0 else (θ1 if t<=z1 else θ2)
        S[t]=S[t-1]+κ*(θ-S[t-1])+σ*rng.normal()
        if rng.random()<λ: S[t]+=rng.normal(μ,s)
        if S[t]<lb: S[t]=lb+0.3*(S[t]-lb)
        Xt=z.iloc[t].values
        S[t]+=0.5*np.mean([Xt@β[m][1:] for m in β])
    return S

def _kalman(x):
    kf=KalmanFilter(dim_x=1,dim_z=1)
    kf.x=[[x[0]]]; kf.F=kf.H=np.eye(1); kf.P*=100; kf.R[:]=1; kf.Q[:]=0.01
    out=[]
    for v in x: kf.predict(); kf.update([[v]]); out.append(kf.x[0,0])
    return np.array(out)

β   = _beta(y_df,m_z); pca = _pca(y_df)
print("\n=== β̂m (macro loading per maturity) ===")
for m, coef in β.items():
    print(f"{m:>5}: {coef[1]:+.4f}")
pca = _pca(y_df)
θ0, θ1, θ2 = pca[:50].mean(), pca[50:150].mean(), pca[150:].mean()
x0  = [0.15,0.015,0.005,0.10,0.05, θ0,θ1,θ2]
bnd = [(1e-4,1),(1e-4,0.05),(0,0.1),(-.5,.5),(1e-4,0.2),(-5,5),(-5,5),(-5,5)]
obj = lambda p: ((_sim(len(pca),pca[0],m_z,β,p)-pca)**2).sum()
ψhat= minimize(obj,x0,bounds=bnd,method='L-BFGS-B',
               options={'maxiter':500}).x

S     = _sim(len(pca),pca[0],m_z,β,ψhat)
S_sm  = _kalman(S)
final = pd.Series(0.7*pca+0.3*S_sm,index=df.index,name='Final SR')

print("\nψ̂:",np.round(ψhat,4))
display(final.head())

ws   = np.arange(0.50, 0.95, 0.05)
rmse = []

for w in ws:
    shadow = w*pca + (1-w)*S_sm
    errs = []
    for m in ylds:
        y_hat = OLS(y_df[m], add_constant(shadow)).fit().fittedvalues
        errs.append(((y_df[m]-y_hat)**2).mean())
    rmse.append(np.sqrt(np.mean(errs))*1e4)

grid = pd.DataFrame({"w": ws, "RMSE_bp": rmse})
print(grid.round(2))

grid.plot(x="w", y="RMSE_bp", marker="o", figsize=(4,3.5))
plt.ylabel("mean cross-section RMSE (bp) (U.S.)"); plt.tight_layout(); plt.show()

plt.figure(figsize=(5,3.5))
plt.bar(range(1,9), PCA().fit(y_df).explained_variance_ratio_)
plt.xlabel('PC'); plt.ylabel('Explained var share')
plt.title('Scree Plot – United States Yield Curve'); plt.tight_layout(); plt.show()

plt.figure(figsize=(11,5))
plt.plot(df.index,final,label='Final SR')
plt.plot(df.index,pca,'--',label='PCA'); plt.grid(); plt.legend()
plt.title('US shadow-rate'); plt.show()

start,end='2009-01-01','2022-08-01'; alpha=0.05
ewma = final.ewm(alpha=alpha,adjust=False).mean().loc[start:end]
wu   = df['SR US'].loc[start:end]

plt.figure(figsize=(11,5))
plt.plot(final.resample('M').mean().loc[start:end],lw=2,label='MAJD')
plt.plot(wu.resample('M').last(),'--',label='WX'); plt.grid(); plt.legend()
plt.ylabel('%'); plt.title('MAJD vs WX (United States)'); plt.show()

plt.figure(figsize=(11,5))
plt.plot(ewma,lw=2,label=f'Final EWMA (α={alpha})')
plt.plot(wu,'--',label='Wu-Xia'); plt.grid(); plt.legend()
plt.ylabel('%'); plt.title('Final EWMA vs Wu-Xia (daily)'); plt.show()

fit_tbl=[]
for m in ylds:
    rmse_f = np.sqrt(((OLS(y_df[m], add_constant(final)).fit()
                       .fittedvalues - y_df[m])**2).mean())*1e4
    rmse_w = np.sqrt(((OLS(y_df[m], add_constant(df['SR US'])).fit()
                       .fittedvalues - y_df[m])**2).mean())*1e4
    fit_tbl.append([m, rmse_f, rmse_w])

fit_df = pd.DataFrame(fit_tbl, columns=['Maturity','RMSE_final','RMSE_WX'])
print("\n=== In-sample cross-section RMSE (bp) ==="); display(fit_df)

TEST_START = '2023-01-01'
train      = final.loc[:TEST_START]
test_idx   = final.loc[TEST_START:].index
n_train    = len(train)

oos_len   = len(test_idx)

ar1 = AutoReg(train, lags=1, old_names=False).fit()

pred_path = ar1.forecast(steps=oos_len)
pred_path.index = test_idx

fc_df = pd.DataFrame({f'h{h}': pred_path.shift(h-1)
                      for h in range(1, 13)},
                     index=test_idx)

oos_tbl=[]
for m in ylds:
    a,b = OLS(y_df[m].loc[:TEST_START], add_constant(train)).fit().params
    act = y_df[m].loc[TEST_START:]
    for h in range(1,13):
        rmse = np.sqrt(((a + b*fc_df[f'h{h}'] - act)**2).mean())*1e4
        oos_tbl.append([m, h, rmse])

oos_df = (pd.DataFrame(oos_tbl, columns=['Maturity','h','RMSE_bp'])
            .pivot(index='Maturity', columns='h', values='RMSE_bp')
            .reindex(ylds))

print("\n=== OOS RMSE (bp) – forecast horizon h = 1…12 months ===")
display(oos_df.round(2))

OUT = OUTPUT_DIR / "shadow_rate_us.csv"
final.to_csv(OUT, index_label="Date")
final.to_excel(OUTPUT_DIR / "shadow_rate_us.xlsx", sheet_name="US SR")

print(f"\n✓ Shadow-rate series written to:  {OUT.resolve()}")
print(f"  Rows written: {len(final):,}")


import warnings, zipfile, numpy as np, pandas as pd, matplotlib.pyplot as plt
from pathlib import Path
from statsmodels.tsa.api import VAR
from tqdm.notebook import trange
warnings.filterwarnings("ignore", category=FutureWarning)

ZIP_LOCAL = find_input_file("Replication codes - Cascaldi-Garcia and Galvao JMCB.zip")

START_M, END_M = "2008-01", "2015-12"
p       = 3
H_IRF   = 36
BOOT    = 500
BLOCK   = 3
ROLL_H  = 1
rng     = np.random.default_rng(0)

WORK_DIR = OUTPUT_DIR / "replication_cgg_mo"; WORK_DIR.mkdir(parents=True, exist_ok=True)
with zipfile.ZipFile(ZIP_LOCAL, "r") as z:
    z.extractall(WORK_DIR)

_mat_candidates = list(WORK_DIR.glob("**/Database_CG-G.mat"))
if not _mat_candidates:
    raise FileNotFoundError("Database_CG-G.mat not found inside the Cascaldi-Garcia & Galvao zip.")
MAT_PATH = _mat_candidates[0]

import scipy.io as sio
mat  = sio.loadmat(MAT_PATH, squeeze_me=True, struct_as_record=False)
qidx = pd.date_range("1975-03-31", periods=mat["RGDP"].shape[0], freq="Q-DEC")

wx_mo = (_prep("Shadow Rates US", ["Date", "SHADOW"])["SHADOW"]
           .astype(float)
           .resample("M").mean()
           .rename("SHADOW"))

sr_daily = final.astype(float).rename("Generated Shadow Rate")
sr_mo = sr_daily.resample("M").mean()

shadow_my_mo = wx_mo.reindex(wx_mo.index.union(sr_mo.index)).copy()
shadow_my_mo.loc[sr_mo.index] = sr_mo
shadow_my_mo.name = "SHADOW"

vars_macro = ["LTFP2","RPCE","EBPm","RVOL1m","JLN1m",
              "RGDP","RINV","RHOURS","RDEF","RSP500"]

df_q = pd.DataFrame({v: mat[v].ravel().astype(float) for v in vars_macro},
                    index=qidx)

macro_mo = (df_q
            .resample("M").last()
            .interpolate(method="linear"))

df_wx = pd.concat([macro_mo, wx_mo], axis=1).loc[START_M:END_M].dropna()
df_my = pd.concat([macro_mo, shadow_my_mo], axis=1).loc[START_M:END_M].dropna()

for df in (df_wx, df_my):
    df[:] = (df - df.mean()) / df.std()

print(f"Months in sample: {len(df_wx)}   ({df_wx.index[0]} → {df_wx.index[-1]})")

res_wx, res_my = VAR(df_wx).fit(p), VAR(df_my).fit(p)
print(f"Wu-Xia VAR({p})  T={res_wx.nobs}   |   Our VAR({p})  T={res_my.nobs}")

def girf(res, tgt="RGDP", sh="SHADOW", H=H_IRF):
    Σ = res.sigma_u.to_numpy()
    j, k = res.names.index(tgt), res.names.index(sh)
    ma = res.ma_rep(maxn=H)
    return np.array([ma[h, j, :].dot(Σ[:, k]) / np.sqrt(Σ[k, k])
                     for h in range(H+1)])

def simulate_var_manual(res, u):
    A, c = res.coefs, res.intercept
    k, T, pL = c.size, u.shape[0], A.shape[0]
    y = np.zeros((T+pL, k)); y[:pL] = res.model.endog[-pL:]
    for t in range(pL, T+pL):
        lag = sum(A[i] @ y[t-i-1] for i in range(pL))
        y[t] = c + lag + u[t-pL]
    return y[pL:]

def block_boot_girf(res, draws=BOOT, H=H_IRF, block=BLOCK):
    resid = res.resid.to_numpy(); T = resid.shape[0]
    out = np.empty((draws, H+1))
    for b in trange(draws, desc="boot", leave=False):
        idx = []
        while len(idx) < T:
            s = rng.integers(0, T-block)
            idx.extend(range(s, s+block))
        u_b = resid[np.array(idx[:T])]
        fit = VAR(pd.DataFrame(simulate_var_manual(res, u_b),
                               columns=res.names)).fit(p)
        out[b] = girf(fit, H=H)
    return np.percentile(out, [5,16,84,95], axis=0)

g_wx, g_my  = girf(res_wx), girf(res_my)
band_wx     = block_boot_girf(res_wx)

m = np.arange(H_IRF+1)/12
plt.plot(m, g_wx,  label="WX")
plt.plot(m, g_my, "--", label="MAJD")
plt.fill_between(m, band_wx[1], band_wx[2], alpha=.15)
plt.fill_between(m, band_wx[0], band_wx[3], alpha=.07)
plt.axhline(0); plt.xlabel("years"); plt.ylabel("% dev")
plt.title("RGDP GIRF (monthly VAR, 1 σ shadow shock)")
plt.legend(); plt.tight_layout(); plt.show()

rmse_wx, rmse_my = [], []
start_roll = max(p + ROLL_H + BLOCK, df_wx.shape[1] + 1)

for t in trange(start_roll, len(df_wx) - ROLL_H, leave=False):
    mod_wx = VAR(df_wx.iloc[:t]).fit(p)
    mod_my = VAR(df_my.iloc[:t]).fit(p)

    f_wx = mod_wx.forecast(df_wx.values[t-p:t], ROLL_H)[-1,
             mod_wx.names.index("RGDP")]
    f_my = mod_my.forecast(df_my.values[t-p:t], ROLL_H)[-1,
             mod_my.names.index("RGDP")]

    y_obs = df_wx["RGDP"].iloc[t]
    rmse_wx.append((y_obs - f_wx)**2)
    rmse_my.append((y_obs - f_my)**2)

d = np.array(rmse_wx) - np.array(rmse_my)
L = BLOCK; M=len(d)
var_d = L * np.var([d[i::L].mean() for i in range(L)], ddof=1) / M
DM = d.mean() / np.sqrt(var_d)
print(f"\nRolling 1-step DM = {DM:+5.2f}   (|stat|>1.96 ⇒ 5 %)")

import numpy as np, pandas as pd
from statsmodels.tsa.api import VAR
from tqdm.notebook import trange

BLOCK = 3
def hac_var(d, L=BLOCK):
    M = len(d)
    if M == 0: return np.nan
    block_means = np.array([d[i::L].mean() for i in range(L)])
    return L * block_means.var(ddof=1) / M

def rolling_msfe(df_wx, df_my, p, h):
    rmse_wx, rmse_my = [], []
    start_roll = max(p + h + BLOCK, df_wx.shape[1] + 1)

    for t in range(start_roll, len(df_wx) - h):
        mod_wx = VAR(df_wx.iloc[:t]).fit(p)
        mod_my = VAR(df_my.iloc[:t]).fit(p)

        f_wx = mod_wx.forecast(df_wx.values[t-p:t], h)[-1,
                mod_wx.names.index("RGDP")]
        f_my = mod_my.forecast(df_my.values[t-p:t], h)[-1,
                mod_my.names.index("RGDP")]
        y_obs = df_wx["RGDP"].iloc[t+h-1]

        rmse_wx.append((y_obs - f_wx)**2)
        rmse_my.append((y_obs - f_my)**2)

    d = np.array(rmse_wx) - np.array(rmse_my)
    DM = d.mean() / np.sqrt(hac_var(d))
    return DM, len(d)

sel = VAR(df_wx).select_order(maxlags=6)
p_aic, p_bic = sel.aic, sel.bic
print(f"AIC-optimal p = {p_aic}   |   BIC-optimal p = {p_bic}")

horizons = [1, 3, 6, 12]
results = []
for crit, p_sel in [("AIC", p_aic), ("BIC", p_bic)]:
    for h in horizons:
        DM, N = rolling_msfe(df_wx, df_my, p_sel, h)
        results.append({"Criterion": crit, "p": p_sel,
                        "h (m)": h, "DM-stat": DM, "N-forecasts": N})

robust_df = pd.DataFrame(results)
print("\n===  Diebold-Mariano (Wu-Xia – Our SR)  ===")
print(robust_df.to_string(index=False,
      formatters={"DM-stat": "{:+.2f}".format}))

robust_df.to_excel(OUTPUT_DIR / "dm_robustness_monthly.xlsx", index=False)
print(f"\nΑποθήκευση: {OUTPUT_DIR / 'dm_robustness_monthly.xlsx'}")

import numpy as np, pandas as pd, matplotlib.pyplot as plt
from statsmodels.tsa.api import VAR
from tqdm.notebook import trange

p_max     = 12
horizons  = [1, 3, 6, 12]
blocks    = [3, 6]
subsplits = [("2008-01", "2011-12"), ("2012-01", "2015-12")]

def hac_var(d, L):
    """fixed-b block variance (Andrews)"""
    if len(d) == 0: return np.nan
    block_means = np.array([d[i::L].mean() for i in range(L)])
    return L * block_means.var(ddof=1) / len(d)

def dm_stat(df_wx, df_my, target, p, h, L):
    rmse_wx, rmse_my = [], []
    start = max(p + h + L, df_wx.shape[1] + 1)
    for t in range(start, len(df_wx) - h):
        m_wx = VAR(df_wx.iloc[:t]).fit(p)
        m_my = VAR(df_my.iloc[:t]).fit(p)
        f_wx = m_wx.forecast(df_wx.values[t-p:t], h)[-1, m_wx.names.index(target)]
        f_my = m_my.forecast(df_my.values[t-p:t], h)[-1, m_my.names.index(target)]
        y_obs = df_wx[target].iloc[t + h - 1]
        rmse_wx.append((y_obs - f_wx)**2)
        rmse_my.append((y_obs - f_my)**2)
    d = np.array(rmse_wx) - np.array(rmse_my)
    DM = d.mean() / np.sqrt(hac_var(d, L))
    return DM, len(d)

target = "RGDP"
dm_grid = {h: [] for h in horizons}

for p in range(1, p_max+1):
    for h in horizons:
        DM, _ = dm_stat(df_wx, df_my, target, p, h, L=3)
        dm_grid[h].append(DM)

plt.figure(figsize=(7,4))
for h in horizons:
    plt.plot(range(1, p_max+1), dm_grid[h], marker="o", label=f"h={h}")
plt.axhline(1.96, color="grey", ls="--", lw=.7)
plt.axhline(-1.96, color="grey", ls="--", lw=.7)
plt.xlabel("VAR lag length p"); plt.ylabel("DM-stat (RGDP)")
plt.title("p-sweep: Diebold-Mariano  (positive ⇒ MAJD better)")
plt.legend(); plt.tight_layout(); plt.show()

print("\n===  DM με HAC block = 6 μήνες  (target: RGDP, p=6) ===")
for h in horizons:
    DM, N = dm_stat(df_wx, df_my, "RGDP", p=6, h=h, L=6)
    print(f"h={h:>2}:  DM={DM:+5.2f}   N={N}")

print("\n===  Sub-sample DM  (p=6, block=3) ===")
for (start, end) in subsplits:
    dfx_w = df_wx.loc[start:end]; dfx_m = df_my.loc[start:end]
    DM, N = dm_stat(dfx_w, dfx_m, "RGDP", p=6, h=1, L=3)
    print(f"{start}–{end}:  DM={DM:+5.2f}   N={N}")

alt_targets = ["RGDP", "RHOURS"]
print("\n===  DM για εναλλακτικά target-vars (p=6, h=1, L=3) ===")
for tgt in alt_targets:
    if tgt not in df_wx.columns:
        print(f"{tgt}:  **δεν υπάρχει στη βάση**")
        continue
    DM, N = dm_stat(df_wx, df_my, tgt, p=6, h=1, L=3)
    print(f"{tgt}:  DM={DM:+5.2f}   N={N}")


import numpy as np, matplotlib.pyplot as plt
from statsmodels.tsa.api import VAR

import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from statsmodels.tsa.api import VAR
import matplotlib.pyplot as plt

wu_xia_us = _prep("Shadow Rates US", ["Date", "WX (United States)"])
wu_xia_us["WX (United States)"] = wu_xia_us["WX (United States)"].astype(float)

shadow_rates_us = final.to_frame("Shadow Rate US")

industrial_prod_us = _prep("Industrial Production US", ["Date", "Ind Prod US"])
cpi_us             = _prep("Inflation Data US", ["Date", "CPI US"])
unemployment_us    = _prep("Unemployment Data US", ["Date", "Unemployment US"])
bond_yields_us_3m  = _prep("3M", ["Date", "Bond Yield US 3M"])
bond_yields_us_10y = _prep("10Y", ["Date", "Bond Yield US 10Y"])

us_data_with_wu_xia = pd.concat([
    wu_xia_us,
    shadow_rates_us.rename(columns={'Shadow Rate US': 'MAJD (United States)'}),
    industrial_prod_us,
    cpi_us,
    unemployment_us,
    bond_yields_us_3m,
    bond_yields_us_10y
], axis=1).dropna()

print(us_data_with_wu_xia.columns)

SHADOW_COLS_US = ['WX (United States)', 'MAJD (United States)']
FACTOR_COLS_US = [c for c in us_data_with_wu_xia.columns if c not in SHADOW_COLS_US]

diff_us = us_data_with_wu_xia.diff().dropna()
diff_us.index = pd.date_range(start=diff_us.index[0], periods=len(diff_us), freq='ME')

standardized_factor_data_us = (diff_us[FACTOR_COLS_US] - diff_us[FACTOR_COLS_US].mean()) / diff_us[FACTOR_COLS_US].std()

pca_us = PCA(n_components=3)
pca_factors_us = pca_us.fit_transform(standardized_factor_data_us)

common_factors_us = pd.DataFrame(
    pca_factors_us,
    index=diff_us.index,
    columns=['Factor1', 'Factor2', 'Factor3']
)
common_factors_us[['Factor1', 'Factor3']] *= -1

def process_data_for_VAR(diff_data, common_factors, shadow_rate_col):
    var_data = pd.concat([common_factors, diff_data[[shadow_rate_col]]], axis=1).dropna()
    return var_data

us_favar_data_your_shadow = process_data_for_VAR(
    diff_us,
    common_factors_us,
    shadow_rate_col='MAJD (United States)'
)

us_favar_data_wu_xia_shadow = process_data_for_VAR(
    diff_us,
    common_factors_us,
    shadow_rate_col='WX (United States)'
)

def fit_VAR_and_analyze(data, lags=2, n_periods=10):
    """
    • Fits a VAR(lags) to the input DataFrame.
    • Prints the full coefficient tables, information criteria,
      and residual-correlation matrix.
    • Returns both the fitted VAR object and its orthogonalised IRF object.
    """
    model     = VAR(data)
    var_model = model.fit(lags)

    print("\n=== VAR({}) estimation results ===".format(lags))
    print(var_model.summary())
    print("\nCorrelation matrix of residuals")
    print(var_model.resid.corr())

    irf = var_model.irf(n_periods)
    return var_model, irf


var_your_shadow, irf_your_shadow = fit_VAR_and_analyze(
    us_favar_data_your_shadow, lags=2, n_periods=10)

var_wu_xia, irf_wu_xia_shadow = fit_VAR_and_analyze(
    us_favar_data_wu_xia_shadow, lags=2, n_periods=10)

print("Impulse Response for the MAJD (United States):")
fig = irf_your_shadow.plot(orth=True)
for ax in fig.axes:
    ax.set_title(ax.get_title().replace("MAJD (United States)", "MAJD (U.S.)"))
plt.show()

print("Impulse Response for WX (United States):")
fig = irf_wu_xia_shadow.plot(orth=True)
for ax in fig.axes:
    ax.set_title(ax.get_title().replace("WX (United States)", "WX (U.S.)"))
plt.show()

def calculate_cumulative_irf(irf_object, n_periods=10):
    cumulative_irf = {}
    irf_matrix = irf_object.orth_irfs[:n_periods]
    response_names = irf_object.model.names

    for i, impulse in enumerate(response_names):
        for j, response in enumerate(response_names):
            response_values = irf_matrix[:, j, i]
            cumulative_irf[f"{response} → {impulse}"] = np.sum(np.abs(response_values))

    return cumulative_irf

cumulative_irf_your_shadow = calculate_cumulative_irf(irf_your_shadow, n_periods=10)
cumulative_irf_wu_xia_shadow = calculate_cumulative_irf(irf_wu_xia_shadow, n_periods=10)

comparison_df = pd.DataFrame({
    "MAJD (United States)": cumulative_irf_your_shadow,
    "WX (United States)": cumulative_irf_wu_xia_shadow
}).T

comparison_df.loc['Difference'] = (
    comparison_df.loc['MAJD (United States)'] - comparison_df.loc['WX (United States)']
)

print("Cumulative Impulse Responses Comparison for US:")
print(comparison_df)

output_file = OUTPUT_DIR / 'cumulative_irf_comparison_us.csv'
comparison_df.to_csv(output_file)
print(f"Comparison data saved as '{output_file}'.")

comparison_df.loc["MAJD (United States)"].plot(kind="bar", figsize=(10, 6), alpha=0.7, label="MAJD")
comparison_df.loc["WX (United States)"].plot(kind="bar", figsize=(10, 6), alpha=0.7, label="WX", color="orange")
plt.title("Cumulative Impulse Responses Comparison (United States)")
plt.xlabel("Response → Impulse")
plt.ylabel("Cumulative IRF")
plt.legend()
plt.tight_layout()
plt.show()


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.api import VAR

shadow_us = final.to_frame("MAJD (United States)")
inflation_us = _prep("Inflation Data US", ["Date", "INF US"])
gdp_us = _prep("Real GDP US", ["Date", "GDP US"])

data_us = pd.concat([shadow_us, inflation_us, gdp_us], axis=1).dropna()

print("\nMerged Data Columns (US):")
print(data_us.columns)

data_us['INF US'] = data_us['INF US'].pct_change() * 100
data_us['GDP US'] = data_us['GDP US'].pct_change() * 100

data_us = data_us.dropna()

print("\nAfter transformations (first few rows):")
print(data_us.head())

def prepare_var_data(df):
    standardized_data = (df - df.mean()) / df.std()
    return standardized_data

var_data_us = prepare_var_data(data_us)

model_us = VAR(var_data_us)
lag_order = 2
results_us = model_us.fit(lag_order)

print("\nVAR Model Summary (US):")
print(results_us.summary())

n_periods = 20
irf_us = results_us.irf(n_periods)

print("\nImpulse Response Functions (US system):")
fig = irf_us.plot(orth=True)
for ax in fig.axes:
    ax.set_title(
        ax.get_title()
        .replace("INF US", "CPI U.S.")
        .replace("GDP US", "Real GDP U.S.")
        .replace("MAJD (United States)", "MAJD (U.S.)")
    )
plt.suptitle('Impulse Response Functions - MAJD (United States)')
plt.show()

def calculate_cumulative_irf(irf_obj, n_periods=10):
    cumulative_irf = {}
    irf_matrix = irf_obj.orth_irfs[:n_periods]
    response_names = irf_obj.model.names

    for i, impulse in enumerate(response_names):
        for j, response in enumerate(response_names):
            cumulative_irf[f"{response} → {impulse}"] = np.sum(np.abs(irf_matrix[:, j, i]))

    return cumulative_irf

cumulative_irf_us = calculate_cumulative_irf(irf_us)
cumulative_irf_us_df = pd.DataFrame(cumulative_irf_us, index=['Cumulative IRF'])

print("\nCumulative IRF Results (US):")
print(cumulative_irf_us_df.T)

cumulative_irf_us_df.T.to_excel(OUTPUT_DIR / 'cumulative_irf_final_shadow_us.xlsx')
print(f"\nCumulative IRF saved to '{OUTPUT_DIR / 'cumulative_irf_final_shadow_us.xlsx'}'.")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.api import VAR

def process_data_for_VAR_alt(diff_data, common_factors, shadow_rate_col):
    var_data = pd.concat([diff_data[[shadow_rate_col]], common_factors], axis=1).dropna()
    return var_data

us_favar_data_alt_your_shadow = process_data_for_VAR_alt(
    diff_us,
    common_factors_us,
    shadow_rate_col='MAJD (United States)'
)

us_favar_data_alt_wu_xia_shadow = process_data_for_VAR_alt(
    diff_us,
    common_factors_us,
    shadow_rate_col='WX (United States)'
)

print("Alternative-order VAR data columns (MAJD):")
print(us_favar_data_alt_your_shadow.columns.tolist())

print("\nAlternative-order VAR data columns (WX):")
print(us_favar_data_alt_wu_xia_shadow.columns.tolist())

var_alt_your_shadow, irf_alt_your_shadow = fit_VAR_and_analyze(
    us_favar_data_alt_your_shadow, lags=2, n_periods=10
)

var_alt_wu_xia, irf_alt_wu_xia_shadow = fit_VAR_and_analyze(
    us_favar_data_alt_wu_xia_shadow, lags=2, n_periods=10
)

print("Impulse Response for the MAJD (U.S.) - Alternative Ordering:")
fig = irf_alt_your_shadow.plot(orth=True)
for ax in fig.axes:
    ax.set_title(
        ax.get_title()
        .replace("MAJD (United States)", "MAJD (U.S.)")
    )
plt.suptitle("Orthogonalised IRFs - MAJD (U.S.) | Alternative Ordering", y=1.02)
plt.tight_layout()
plt.show()

print("Impulse Response for WX (U.S.) - Alternative Ordering:")
fig = irf_alt_wu_xia_shadow.plot(orth=True)
for ax in fig.axes:
    ax.set_title(
        ax.get_title()
        .replace("WX (United States)", "WX (U.S.)")
    )
plt.suptitle("Orthogonalised IRFs - WX (U.S.) | Alternative Ordering", y=1.02)
plt.tight_layout()
plt.show()

cumulative_irf_alt_your_shadow = calculate_cumulative_irf(irf_alt_your_shadow, n_periods=10)
cumulative_irf_alt_wu_xia_shadow = calculate_cumulative_irf(irf_alt_wu_xia_shadow, n_periods=10)

comparison_alt_df = pd.DataFrame({
    "MAJD (United States) - AltOrder": cumulative_irf_alt_your_shadow,
    "WX (United States) - AltOrder": cumulative_irf_alt_wu_xia_shadow
}).T

comparison_alt_df.loc["Difference"] = (
    comparison_alt_df.loc["MAJD (United States) - AltOrder"]
    - comparison_alt_df.loc["WX (United States) - AltOrder"]
)

print("\nCumulative IRF Comparison for U.S. (Alternative Ordering):")
print(comparison_alt_df)

comparison_alt_df.to_csv(OUTPUT_DIR / "cumulative_irf_comparison_us_alt_order.csv")
print(f"\nSaved: {OUTPUT_DIR / 'cumulative_irf_comparison_us_alt_order.csv'}")

def shadow_shock_only(irf_object, shadow_name, n_periods=10):
    irf_matrix = irf_object.orth_irfs[:n_periods]
    names = irf_object.model.names
    shock_idx = names.index(shadow_name)

    out = {}
    for j, response in enumerate(names):
        out[f"{response} ← shock in {shadow_name}"] = np.sum(np.abs(irf_matrix[:, j, shock_idx]))
    return pd.Series(out)

shadow_only_alt_majd = shadow_shock_only(
    irf_alt_your_shadow,
    shadow_name="MAJD (United States)",
    n_periods=10
)

shadow_only_alt_wx = shadow_shock_only(
    irf_alt_wu_xia_shadow,
    shadow_name="WX (United States)",
    n_periods=10
)

shadow_only_alt_df = pd.concat(
    [shadow_only_alt_majd, shadow_only_alt_wx],
    axis=1
)
shadow_only_alt_df.columns = ["MAJD AltOrder", "WX AltOrder"]
shadow_only_alt_df["Difference"] = shadow_only_alt_df["MAJD AltOrder"] - shadow_only_alt_df["WX AltOrder"]

print("\nCumulative responses to SHADOW-RATE shock only (Alternative Ordering):")
print(shadow_only_alt_df)

shadow_only_alt_df.to_csv(OUTPUT_DIR / "shadow_shock_only_us_alt_order.csv")
print(f"\nSaved: {OUTPUT_DIR / 'shadow_shock_only_us_alt_order.csv'}")
```
